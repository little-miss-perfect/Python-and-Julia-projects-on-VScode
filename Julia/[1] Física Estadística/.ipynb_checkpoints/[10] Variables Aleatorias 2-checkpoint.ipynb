# Variables Aleatorias

La clase pasada vimos la definición de una variable aleatoria como una función del espacio de muestreo a los reales. Pero en el fondo no nos importa qué función sea, pues lo que queremos es medir la probabilidad de que esa función nos arroje algún valor o conjunto de valores. Es decir, nos asumimos desconocedores de cómo se asignan esos valores, no nos importa tanto la función, sólo nos importa con qué probabilidad arroja cada valor. 

En ese sentido, en probabilidad nos basta con tener el espacio de muestreo (para saber de qué estamos hablando) y la distribución de probabilidad. 

En probabilidad, nos basta con $\Omega$ y $\rho_X(x)$ para estudiar las propiedades de la variable aleatoria $X$. 

Al final, ya vimos que para una variable aleatoria discreta:

$P(X \in K) = \sum_{x \in K} p_{X}(x)$

Ahora bien, en muchos problemas la probabiildad que buscamos es de la forma: 

$F_X(a) = P(X\le a)$

en variables discretas:

$ = \sum_{x\le a} p_X(x)$

Por ejemplo, si quiero saber cuál es la probabilidad de reprobar este curso, es la probabilidad de que mi calificación sea menor o igual a 5. 

Esta función se llama la función cumulativa. Lo importante de esta función es que nos va a ayudar a definir el equivalente a la probabilidad de masa en las variables continuas, pues aunque la probabilidad de masa es $0$ para todos los valores de $x$, la función cumulativa no es necesariamente $0$. 

Primero mostremos algunas propiedades de la función cumulativa. 

### Propiedades de la función cumulativa

$1\ge F_X(a) \ge 0$


Lo primero que vemos obvio es que la función cumulativa es una función no decreciente de $x$, pues es la probabilidad de que $X$ sea menor que $x$. Además, en el límite donde $x$ tiende a $-\infty$, la función cumulativa es $0$ y en el límite donde $x\rightarrow \infty$, la función cumulativa es 1. 

También notemos que la probabilidad de $P(X\le x) = P(X<x) + P(X = x)$, que en el caso de las variables continuas, $P(X = x) = 0$. 

Pero además se va a cumplir que:

$P(a < X\le b) = F_X(b) - F_X(a)$

Para demostrarlo sólo tenemos que notar que $\{X\le b\} = \{X \le a\}\cup \{a < X \le b\}$ y que los dos conjuntos que unimos son conjuntos disjuntos. De ahí en fuera es sólo la definición. 

Además, por lo mencionado arriba, podemos cambiar los $\le$ por $<$ y visebersa en el caso de las variables continuas. Si no son continuas, no podemos, pero se puede jugar con las probabilidades de $a$ y $b$ para corregir la fórmula. 

Ahora bien, hacemos el truco de cálculo: 

$P(x < X \le x+\Delta x) = F_X(x+\Delta x) - F_X(x)$

Si $F_X$ es diferenciable, entonces esta función la podemos aproximar por: 

$P(x < X \le x+\Delta x) = F_X(x+\Delta x) - F_X(x) \approx F'_X(x)\Delta x$

Usando el teorema fundamental del cálculo llegamos a que: 

$P(a  < X \le b) = \int_a^b F_X'(x) dx = \int_a^b \rho_X(x) dx$

donde $\rho_X$ es la densidad de probabilidad, que es el equivalente la probabilidad de masa. También llamaremos a $\rho_X$ la distribución de $X$ en el caso de las variables continuas. En ese caso, las desigualdades estrictas y no estrictas son intercambiables. 

Por otro lado, si $X$ es una variable discreta, $F_X$ no es diferenciable, porque ni siquiera es continua, pero se puede aproximar por una suma de funciones paso de Heaviside: 

Función de Heaviside: 

$H(x)=
\begin{cases}
    0 & \mathrm{si} & x < 0 \\
    1 & \mathrm{si} & x \ge 0
    \end{cases}$

Para variables discretas: 

$F_X(x) = \sum_i a_i H(x-x_i)$ 

con $\sum_i a_i = 1$
    
    

La función paso no es diferenciable en los reales, pero lo es en los reales extendidos. En este caso, la derivada de $F_X$ es una delta de dirac. Así que: 

$$P(a  < X \le b) = \int_a^b \rho_X(x') dx' = \int_a^b F_X'(x') dx' = \sum_i \int_a^b a_i\delta(x'-x_i)dx'$$ 

Y la $\delta(x)$ de dirac, es una distribución normal con parámetros $\mu = 0$ y $\sigma \rightarrow 0$

$\delta(x) = \lim_{\sigma \rightarrow 0} \frac{1}{\sqrt{2\pi}} \exp\Big(\frac{-x^2}{2\sigma^2}\Big)$

Ahora, si hacemos $a = x$ y tender a $b = x$ tendremos la probabilidad de $X = x$, que es la probabilidad de masa, pero esto es la derivada de la integral multiplicada por $d x$, es decir $\rho_X(x) dx$ y por eso decimos que la densidad de probabilidad es el equivalente a la probabilidad de masa. 

Notemos además que dada esta definición de la densidad de probabilidad, la integral de $-\infty$ a $\infty$ de la densidad de probabilidad de cualquier variable aleatoria es igual a $1$, ya que esto es $P(-\infty < X\le \infty) = 1$.  

$$\int_{-\infty}^\infty \rho_X(x) dx = 1$$

Si tenemos la distribución de probabilidad, tenemos prácticamente todo de nuestra variable aleatoria. La pregunta es ¿Cómo le hacemos para obtener esta distribución de probabilidad? 

La clase pasada ya mencioné que podemos obtener experimentalmente la probabilidad de masa haciendo un histograma. Y una variable aleatoria continua siempre podemos aproximarla por una discreta. Pero también podemos partir de suposiciones donde fijemos una variable aleatoria vía argumentos heurísticos, por ejemplo con una gaussiana, equiprobabilidad, una distribución de Poisson, etc. Y después, operar con esa variable aleatoria para obtener una nueva, más complicada.  

La pregunta natural es, si yo conozco $\rho_X$ para una variable $X$, ¿puedo calcular $\rho_Y$ para una variable $Y = f(X)$?




Si $f$ es una función biyectiva y $Y = f(X)$, entonces: 

$p_Y(y) = P(Y = y) = P(f(X) = y) = P(X = f^{-1}(y)) = p_X(f^{-1}(y)) = p_X(x)$


Esto claro, si la función es biyectiva. Si no lo fuera entonces habría para cada valor $y$ varios posibles valores $x_i$ que cumplieran con que $f(x_i) = y$. Por lo tanto la probabilidad de masa de $Y$ se incrementaría en $y$, pues cada uno de los valores de $x_i$ contribuirían $p_X(x_i)$ a la probabilidad de masa de $Y$ en $y$. Es decir: 

$p_Y(y) = P(Y = y)  = P(f(X) = y) = \sum_{f(x) = y} p_X(x) $

Por ejemplo, digamos que $f(x) = x^2$, y $p_X(x)$ es la probabilidad de $1/101$ para todos los números enteros entre $-50$ y $50$. Entonces:

$p_Y(y) = \sum_{x^2 = y} p_X(x) = \begin{cases}
    \frac{2}{101} & \mathrm{si} & x^2 \ne 0 \\
    \frac{1}{101} & \mathrm{si} & x = 0
    \end{cases}$

```julia
using Plots, StatsBase, LinearAlgebra
n = 1_000_000
X = [rand(-50:50) for i in 1:n]

Y = [(rand(-50:50))^2 for i in 1:n]

x = -50:50 .-0.5
y = [i^2.0.-0.5 for i in 0:51]

hx = fit(Histogram, X, x)
hx = normalize(hx, mode=:probability)
bar(x[1:end-1].+0.5,hx.weights, key = false, lw = 0)
hline!([1/101])
```

```julia
function py(y)
    if y < 1
        return 1/101
    else
        return 2/101
    end
end
N = 10000
hy = fit(Histogram, Y, y)
hy = normalize(hy, mode=:probability)

bar([(y[i]+y[i+1])/2 for i in 1:length(y)-1].+0.5,hy.weights, key = false, bar_width = [(y[i]-y[i+1])*0.9 for i in 1:length(y)-1], lw = 0)
plot!((0.1:0.1:10) ∪ (11:5000), py.((0.1:0.1:10) ∪ (11:5000)))
plot!(xlim = (0.1,N), xscale = :log)
```

Podemos hacer ahora el mismo truco para el caso continuo.  

Por ejemplo, digamos que queremos obtener ahora la distribución de una variable uniforme en el intervalo $[-1,1]$ al cuadrado. Entonces: 

$\rho_X(x) = 1/2$ si $x \in [-1,1]$

$F_Y(y) = P(Y\le y) = P(X^2 \le y) = P(-\sqrt{y} \le X \le \sqrt{y}) = F_X(\sqrt{y})- F_X(-\sqrt{y})$

por lo tanto 

$\rho_Y(y) = F'_Y(y) = \frac{1}{2\sqrt{y}} (F'_X(\sqrt{y}) + F'_X(-\sqrt{y})) = \frac{1}{2\sqrt{y}}(\rho_X(\sqrt{y})+\rho_X(-\sqrt{y})) = \frac{1}{2\sqrt{y}} $

```julia
n = 100_000
X = [1-2*rand() for i in 1:n]

Y = [(1-2*rand())^2 for i in 1:n];
```

```julia
x = -1:0.001:1
y = 0:0.001:1
bins = 1000
histogram(Y, nbins = bins)
plot!(y, n/bins*0.5 ./sqrt.(y), lw  = 4)
```

Con esto ya tenemos como calcular todo una gama de distribuciones de probabilidad, aunque no siempre es obvio cómo obtener el conjunto de todos los valores $x$, tales que $f(x) = y$. 

Dicho esto, pasemos a estudiar el valor esperado de una variable aleatoria. 

## Valor esperado

Para esto quiero recordarles lo que ya vimos, que el valor esperado de una función real en un intervalo es:

$\langle g(x) \rangle_a^b = \frac{1}{b-a} \int_a^b g(x)dx$

$\frac{1}{b-a}\int_a^b f(x)dx = lim_{n \uparrow} \Sigma_{k = -\infty}^\infty \frac{(k-1)}{2^n} P(\{a\le x\le b: (k-1) 2^{-n} \le f(x) < k 2^{-n} \})$

Pero esto carece de significado en variables aleatorias porque las variables aleatorias van del espacio de muestreo a los reales. Así que no puede significar lo mismo "el promedio de una función", que "el promedio de una variable aleatoria". 

Ahora bien, la distribución de probabilidad es una función que ya está, por así decirlo, dividida entre el invervalo, porque su integral debe dar 1 en todo el espacio.  

¿Qué significa el valor esperado de una variable aleatoria? Sería algo así como el valor promedio que arroje esa variable. Una variable aleatoria puede en principio arrojar cualquier número real $x$, pero lo hace con una probabilidad (de masa o densidad ) diferente en cada caso. Así que el valor promedio será la suma de la probabiidad de obtener $x$ multiplicado por $x$, es decir: 

$$\langle X \rangle =  \lim_{\Delta x \rightarrow 0}\sum_x xP(x < X \le x+\Delta x)  = \int_{-\infty}^\infty x\rho_X(x)dx $$

O bien su versión con la sumas, pero que como ya dijimos, podemos obnter de sustituir en la densidad de probabilidad una suma de deltas de dirac.

Por ejemplo: 

**Ejemplo** 

¿cuál es el valor esperado de una variable aleatoria con una distribución binomial?

$P_X(m) = \binom{n}{m} p^m (1-p)^{n-m}$

$\Rightarrow \langle X \rangle = \sum_m m\binom{n}{m} p^m (1-p)^{n-m} = \sum_{m= 1}^n \frac{ n!}{(m-1)! (n-m)!} p^m (1-p)^{n-m} $



$= np\sum_{m= 1}^n \frac{ (n-1)!}{(m-1)! (n-m)!} p^{m-1} (1-p)^{n-m} = np \sum_{m = 1}^{n} \binom{n-1}{m-1} p^{m-1} (1-p)^{n-m}$



$= np \sum_{j = 0}^{n-1} \binom{n-1}{j} p^{j} (1-p)^{(n-1)-j} = np (p + (1-p))^{n-1} = np
$

¿Cuál es el valor esperado de una variable aleatoria con distribución normal? 

$\rho_X(x) = \frac{1}{\sqrt{2\pi}\sigma} \exp\Big(-\frac{(x-\mu)^2}{2\sigma^2}\Big)$

$\langle X \rangle = \int_{-\infty}^\infty x \frac{1}{\sqrt{2\pi}\sigma} \exp\Big(-\frac{(x-\mu)^2}{2\sigma^2}\Big) dx$

haciendo cambio de variable $x= \sigma t+\mu$, $dx = \sigma dt$


$\langle X \rangle = \frac{\sigma}{\sqrt{2\pi}\sigma} \int_{-\infty}^\infty (\sigma t+\mu) \exp\Big(-\frac{t^2}{2}\Big) dt$

$ = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^\infty \mu \exp\Big(-\frac{t^2}{2}\Big) dt + \frac{1}{\sqrt{2\pi}} \int_{-\infty}^\infty \sigma t \exp\Big(-\frac{t^2}{2}\Big) dt$


La primera integral es de una función par, la segunda es de una función impar, así que la segunda es 0 y la primera ya la habíamos integrado y es $\sqrt{2\pi}$ por $\mu$. Por lo tanto: 

$\langle X \rangle = \mu$

Para lo que sigue necesitamos definir la densidad de probabilidad conjunta, que es la probabilidad de que ocurran dos variables aleatorias al mismo tiempo como $\int_a^b \int_c^d \rho_{X,Y}(x,y) = P(a\le X\le b, c\le Y\le d)$. Por lo tanto, se tendrá que: 

$\int_{-\infty}^\infty \rho_{X_1,X_2, \dots, X_i,\dots, X_n}(x_1,x_2,\dots, x_i,\dots, x_n)dx_i = \rho_{X_1,X_2, \dots, X_n}(x_1,x_2,\dots, x_n)$ 

En particular, si tenemos una variable aleatoria compuesta de 2 variables aleatorias, tendremos que: 

$\int_{-\infty}^\infty \rho_{X_1,X_2}(x_1,x_2)dx_i = \rho_{X_1}(x_1)$

Ahora bien, para calcular el valor esperado de $Z(X_1, X_2, \dots, X_n)$, podemos utilizar $\rho_Z$ o  $\rho_{X_1,X_2, \dots, X_n}$. En el primer caso integramos sobre todos los posibles valores de $Z$, mientras que en el segundo caso integramos sobre cada uno de los valores de $X_i$. Es equivalente, la integración, aunque no son equivalentes las $\rho$'s. 

Aplicando esto a la suma de dos variables aleatorias tenemos que: 

$$\langle X+Y \rangle = \int_{-\infty}^{\infty}\int_{-\infty}^{\infty} (x+y)\rho_{X,Y}(x,y)dxdy = \int_{-\infty}^{\infty} x\rho_{X}(x)dx +\int_{-\infty}^{\infty} y\rho_{Y}(y)dy = \langle X \rangle + \langle Y \rangle$$ 

Es decir, el valor esperado es un operador lineal. 

Por otro lado, podemos calcular la probabilidad de una función aplicada a una variable aleatoria de la siguiente forma:

$$\langle f(X) \rangle = \int_{-\infty}^\infty f(x) \rho_X(x) dx$$  

Para demostrarlo primero usamos la función indicatriz $I_A(x)$, que es igual a 1 si $x \in A$ y $0$ si no. El valor esperado de esta variable aleatoria es $\langle I_A \rangle = \int_{-\infty}^\infty x I_A(x)dx  = 0 p(0) + 1 p(1) = P(A)$. Entonces, si $I_A$ es la función indicatriz de un segmento $A$, $\langle I_A(X) \rangle$ será igual a la probabilidad de que la variable aleatoria $X$ caiga en el subconjunto $A$ de los reales. Pero esto esigual a la integral de la densidad de probabilidad de $X$ sobre el subconjunto $A$. Es decir: 

$$\langle I_A(X) \rangle = P(X \in A) = \int_A \rho_X(x) dx = \int_{-\infty}^\infty I_A(x) \rho_X(x) dx $$

El siguiente paso es notar que cualquier función se puede aproximar tan bien como querramos como una suma infinita de funciones indicatriz y utilizar que el valor esperado de la suma de variables aleatorias es la suma de los valores esperados. 

Por lo tanto, en el límite de esa aproximación se cumple el teorema que queríamos mostrar.

Ahora revisemos qué pasa con el valor esperado del producto de dos variables aleatorias. Pero antes necesitamos la versioón de la probabilidad condicional, en versión densidad, la densidad de probabilidad condicional como $\rho_{Y|X}(y|x) = \frac{\rho_{X,Y}(x,y)}{\rho_X(x)}$. 

Esta definición es equivalente a la que vimos de probabilidad condicional en su versión densidad. Arriaba la proabilidad de la intersección, abajo la probabilidad del evento que condiciona. La utilizaremos más adelante para

Ahora sí: 

$\langle XY\rangle = \int_{-\infty}^{\infty}\int_{-\infty}^{\infty} xy\rho_{XY}(x,y) dx dy$ 

que si $X$ y $Y$ son independientes: 

$= \int_{-\infty}^{\infty}\int_{-\infty}^{\infty} xy\rho_{X}(x)\rho_{Y}(y) dx dy = \langle X\rangle\langle Y\rangle$

Ese último paso se debe a que si las variables son independientes, pues la probabilidad del producto es una probabilidad conjunta, que es la densidad de probabilidad de $Y$ dado $X= x$ por la densidad de probabilidad de $X$, pero si son independientes, la probabilidad condicional es simplemente la probabilidad de $Y$, entonces la probabilidad del conjunto es el producto de las probabilidades. Y la densidad de probabilidad es justamente la probabilidad condicional a que ocurra X dado Y. 

Ok, hasta aquí, creo que no hemos dicho nada tan sorprendente y probablemente, si llevaron cuántica, usaron ya estas fórmulas aunque no las hayan demostrado. Ya las demostramos. 

Ahora bien, pensemos en el caminante aleatorio, donde tiramos volados y en cada sol nos movemos a la derecha y cada águila a la izquierda. Si la probabilidad de que salga sol o águila es la misma, sabemos que en promedio el caminante no se moverá. Pero en general no se encontrará cercano a 0 la posición después de $n$ tiros

```julia
function caminante_aleatorio1D(x,n)
    posiciones = [x]
    while n>0
        push!(posiciones, posiciones[end]+rand([-1,1]))
        n -= 1
    end
    return posiciones
end

function caminante_aleatorio1D_final(x,n)
    while n>0
        x += rand([-1,1])
        n -= 1
    end
    return x
end
```

```julia
using Plots, LaTeXStrings
plot(caminante_aleatorio1D(0, 1000), key = false)
hline!([0], xlabel = L"t", ylabel = L"x")
```

```julia
caminantes = [caminante_aleatorio1D_final(0, 1000) for i in 1:1000]
histogram(caminantes, nbins = 100)
```

Así que sí, en promedio llega al 0, pero en general no y nos gustaría dar una medida de qué tanto no llega al 0. Podríamos sólo contar cuantas veces llega del total, pero también nos interesa que esa medida nos diga qué tanto se aleja. Esa medida es la varianza

## Varianza

La idea es medir en vez del valor promedio, el promedio de la distancia de la variable al valor promedio. Es decir, el promedio de $\sqrt{(X-\mu)^2}$.  Pero esta cantidad tiene un $\sqrt{}$ que es molesto (es difícil derivar, difícil integrar), así que preferimos sacar el promedio del cuadrado de esta distancia, es decir, el promedio de $(X-\mu)^2$. Ese promedio es la varianza: 

$\sigma^2  = \langle (X-\mu)^2 \rangle = \langle X^2 - 2\mu X + \mu^2 \rangle = \langle X^2\rangle - 2 \langle X \rangle \langle X \rangle + \langle X \rangle^2 =    \langle X^2\rangle -  \langle X \rangle ^2 $

De esta formulita ya hablamos del valor esperado, así que lo único nuevo es el valor esperado del cuadrado de la variable. Ese es el segundo momento. 

Explícitamente podemos calcular la varianza como: 

$$\sigma^2 = \int_{-\infty}^\infty x^2 \rho_X(x) dx - \Big(\int_{-\infty}^\infty x \rho_X(x) dx\Big)^2$$

La varianza así definida es una buena definición de qué tan rápido aproximamos el promedio. 

$$\sigma^2(X) = \int_{-\infty}^\infty (x-\mu_X)^2 \rho_X(x) dx \ge \int_{|x-\mu_x|\ge t} (x-\mu_X)^2 \rho_X(x)dx \ge t^2 \int_{|x-\mu_x|\ge t} \rho_X(x)dx = t^2 P(|X-\mu_X|\ge t)$$

$$ \Rightarrow P(|X-\mu_X|\ge t) \le \frac{\sigma^2(X) }{t^2}$$

Es decir, la varianza realmente nos mide qué tan fácil es medir el promedio de una variable aleatoria. 

Ahora por supuesto, nos preguntamos si podemos, como con el valor esperado, abrir las sumas. Quisiéramos obtener la varianza de la suma de dos variables aleatorias. Tenemos entonces: 

$\sigma^2 (X + Y) = \langle (X+Y)^2\rangle - \langle X+Y\rangle^2 = \langle X^2 \rangle + \langle Y^2 \rangle - \langle X\rangle^2 -\langle Y^2\rangle + 2(\langle XY\rangle - \langle X\rangle\langle Y \rangle) = \sigma^2 (X) + \sigma^2(Y) + 2\langle (X-\mu_X)(Y-\mu_Y)\rangle$

Esta vez no es un operador lineal, pero "casi", falla sólo por un término. Ese término es la covarianza. 

## Covarianza

$Cov(X, Y) = \langle (X-\mu_X)(Y-\mu_Y)\rangle$

Si $X = Y$, entonces la covarianza se vuelve simplemente la varianza. De ahí en fuera, suena a una cantidad sin importancia. Pero analicemos un poco el significado de la covarianza. 

Si $X$ y $Y$ son independientes, entonces la covarianza es el producto de los valores esperados menos los valores esperados, que es 0. Es decir, de alguna forma, la covarianza mide qué tan relacionados están $X$ y $Y$, pues si son la misma variable (muy correlacionados), entonces la covarianza es la varianza. Si son totalmente independientes, entonces la covarianza es 0. 

La covarianza puede también ser $0$ aunque no sean independientes, pues puede pasar que el valor esperado del producto de las variables aleatorias sea igual al producto de los valores esperados sin que sean variables independientes.  

Además, la covarianza permite valores negativos, así que puede haber "anti-correlación".

Por otro lado, es un poco molesto que los valores oscilen entre dos números que dependen de la varianza de $X$ y $Y$, porque entonces es difícil usarlo como una medida de la correlación. Por ello conviene normalizar. Hay varias opciones para normalizar. Por ejemplo, se podría dividir entre el máximo de las varianzas, pero eso hace asimétrica la función. Una mejor opción es dividir entre el producto de las desviaciones estandar. 

Con esta normalización la covarianza siempre tiene valores entre $-1$ y $1$ y le llamamos correlación. 

## Correlación

$$Corr(X,Y) = \frac{Cov(X,Y)}{\sqrt{\sigma^2(X) \sigma^2(Y)}}$$

La correlación será una cantidad bastante importante, no sólo en física estadística, sino en su vida si hacen investigación o se dedican al análisis de datos. Medir correlaciones entre diferentes variables les ayudará a determinar si están relacionadas. Si tienen una buena hipótesis sobre la relación entre 2 variables, pueden medir su correlación para darle más fuerza a la hipótesis. 

Por ejemplo, estoy trabajando con un médico que tenía la hipótesis de que una bacteria estomacal estaba relacionada a la época de lluvias, que de alguna forma, el mecanismo para dispersar la bacteria era mediante el agua de lluvia. Entonces revisamos la correlación con varias variables climatológicas. La temperatura, contaminación del aire, milímetros de lluvia, humedad relativa, etc. Con la variable de número de casos registrados. La correlación más alta no fue con la lluvia, sino con la temperatura. Esto nos llevó a investigar sobre el ciclo de reproducción de la bacteria. Descubrimos que efectivamente en los rangos de temperatura de México, la probabilidad de reproducción incrementa con la temperatura de forma similar a como incrementa el número de casos (lo que nos lleva al problema de la bacteria que se reproduce cada minuto con probabilidad $r$, para temperaturas suficientemente bajas, efectivamente, las colonias de bacterias tendían a reducirse aunque sí había reproducción).

Otro ejemplo, están en una empresa que vende varios productos. Tienen los datos de los clientes, información sobre su situación económica, dónde vive, género, etc. y por supuest, saben qué cosas compran. Una información muy valiosa para la empresa es saber la correlación entre cada variable del cliente con el producto que compra. Así pueden saber a qué sector dirigir la publicidad en cada caso. Si hablamos de una empresa grande, como Liverpool, Wallkmark, etc, encontrar correctamente las correlaciones tiene valor de cientos o miles de millones. Y por eso se ha vuelto popular el análisis de datos en los últimos años. Por supuesto, ahora hay cientos de empresas que se dedican sólo a hacer ese análisis para las grandes y medianas empresas. Tienen varias técnicas, pero lo princial es justamente medir la correlación. O sea, aprender bien esta formulita les puede dar trabajo en el futuro. 

Finalmente, a nosotros en particular nos interesa para estudiar transiciones de fase. En un cristal, la correlación entre la posición de una partícula y otra lejana es muy alta. En un fluido esa correlación decae muy rápidamente con la distancia. Así que la correlación en la posición de las partículas en un material nos dice mucho sobre la fase en la que está y estudiar cómo evoluciona la correlación al cambiar las variables termodinámicas nos peude decir mucho sobre la transición de fase.
