## Historia

### Revoluciones científicas

Es difícil definir una revolución científica. En toda la historia de la humanidad ha habido avances importantes. Incluso en el oscurantismo de la edad media hubo varios avances en todas las áreas de la ciencia. Hubo retrocesos importantes también y por eso le llamamos oscurantismo, pero siempre hubo gente lista que tuvo grandes ideas. Sin embargo sí ha habido momentos clave, a partir de los cuales hubo un derrame de nuevas ideas. De la edad media para atrás, los descubrimientos solían quedarse en la cultura donde salían, así que es difícil hablar de una revolución científica, porque cada cultura tuvo sus propios avances. Hablamos mucho de los griegos, que definitivamente aportaron mucho, pero en realidad al mismo tiempo los árabes, chinos, indios, persas, incas, mayas, etc., tenían desarrollos similares, simplemente los griegos penetraron la cultura occidental que fue la que dominó gran parte del mundo. Y ahí está parte del secreto de las revoluciones científicas después de la edad media. Occidente empezó a tener control de todo el mundo, así que lo que se discutía en Europa se exportaba a todo el mundo. Antes de eso como dije, los avances se daban de forma aislada y paralela. A veces hay resultados que una cultura los tiene un año, 100 años después otra cultura los tiene también, 100 años después otra. Hay hasta un milenio de distancia entre cuando una cultura descubrió una cosa y cuándo lo descubrió otra cultura y a veces la misma cultura lo descubre un año y 100 años después otra vez. Eso se termina cuando Europa conquista casi todo el mundo. Además de conquistar, conectó. No digo que haya sido algo bueno esas conquistas, Europa sometía al resto del mundo y eso no era presisamente bueno, pero en términos científicos hizo una conexión. Una muy injusta, porque no dejaba al resto del mundo expresarse, así que cuando algo surgía en el resto del mundo, era callado o robado. Pero es verdad que eso hizo que el mundo más o menos se homogeneizara en conocimiento. La homogeneización también implicó que hubiera 2 tipos de "descubrimientos científicos", unos que eran resultado de los anteriores, una especie de continuación y los que implicaron un rompimiento de paradigmas, que crearon nuevas teorías. Esto ocurría antes, pero sin que hubiera homogeneización, los cambios de paradigma no eran tan radicales. Para que haya cambios radicales, tiene que haber teorías bien establecidas y eso lleva tiempo de muchas personas. 

Creo que la primera gran revolución la inició Coopérnico y Galileo y seguido de él Kepler, lo que llevó a Newton a proponer sus leyes de la física y con eso no sólo se desarrolló la mecánica, se desarrolló el cálculo y un montón de herramientas importantes. Los trabajos de Galileo, Kepler y Newton despertaron el interés de muchos en Europa (y el mundo), que llevó en verdad a una serie de avances en matemáticas y física. Esto ocurrió entre 1650 y 1730 más o menos. 

Creo que la siguiente gran revolución en la física se dio por dos frentes, la termodinámica y el electromagnetismo gracias a Carnot, Clausius, Maxwell, Faraday, etc. Esto en los 1800's. Más o menos 100 años después de la primera gran revolución. 

Justamente fue en esta época donde se empezó a cocinar la siguiente revolución, que tardó menos de 50 años en llegar. Por supuesto, me refiero a finales del siglo antepasado y principios del siglo pasado, cuando primero se creo la física estadística con las ideas de Boltzmann y muy poco después se creó la mecánica cuántica y la relatividad. Las 3 teorías resultado de las inconsistencias encontradas en el electromagnetismo y la termodinámica. La historia comienza con Gustav Kirchoff, el mismo de las leyes de Kirchoff que ven en electrónica. 

- Gustav Kirchhoff

<img src="kirchhoff.png" alt="Kirchhoff" width="300" height="400">

En 1860 Kirchhoff escribió un largo artículo titulado (en alemán) "Sobre la relación entre la potencia de radiación
y absorción de luz y calor de diferentes cuerpos". Aquí les pongo una re-impresión en inglés para philosophical magazine. 

<img src="Kirchhoff_paper.png" alt="Kirchhoff" width="400" height="600">


En el artículo Kirchhoff discute qué sucede cuando uno pone un cuerpo que radía luz en un baño térmico. Ya para su época se sabía que tanto la luz como el calor eran formas de energía y que si se encontraba un cuerpo en equilibrio térmico, entonces debería de estar radiando tanta energía como recibiera. De alguna forma se empezaba a jugar con la idea de que hubiera luz "no visible" que representara el calor radiado. Aún no se demostraba esto, pero ya se tenía en mente. Entonces, si el cuerpo estaba en equilibrio térmico, debía radiar tanta "luz" (visible y no visible), como la que recibiera. 

Para entonces él mismo y un varios químicos (Bunsen entre ellos) habían estudiado qué sucedía cuando se calentaba un cuerpo como el sodio y habían descubierto los espectros de emisión y absorción. La luz emitida por esos cuerpos al calentarse era de una serie de frecuencias concretas y habían notado que eran las mismas frecuencias que absorvían si se les iluminaba con luz blanca, incluyo ya habían encontrado los espectros de absorción del sol y lo habían relacionado con sus componentes. 


<img src="espectro_emisión_absorcion.png" alt="" width="400" height="600">


Esto por supuesto fue de mucha importancia, porque generaba huellas dactilares de los diferentes compuestos, lo que más tarde sirvió para determinar de qué elementos está compuesto un material. Fue la primera herramienta de la espectroscopía. Creo que si no hubiera venido lo de después, esto tarde o temprano hubiera llevado a una teoría cuántica, pero quizá hubiera tardado bastante más. Kirchhoff notó que para estar en equilibrio térmico debería haber la misma energía emitida que absorvida todo el tiempo. Si se tratara de un solo color eso significaría que la proporción de luz emitida y absorvida sería una constante. La pregunta natural es ¿Eso se mantiene cuando se tienen varios colores? Es decir, ¿hay una razón constante entre la luz emitida y absorvida para cada color? La energía emitida y absorvida debe de ser la misma, pero eso no nos dice nada sobre como se reparten las intensidades y en principio podría depender de la naturaleza interna del material, cómo procesa la luz. Recibe una cantidad de luz, hace algo con esa luz y luego la emite quizá en otra frecuencia con más o menos intensidad, pero de tal forma que se mantenga la cantidad de energía. 

Para estudiar este caso él pensó en un objeto que imaginario, que no reflejara nada de luz, no trasmitiera nada de luz (fuera totalmente opaco), que absorviera toda la luz y además que fuera infinitamente delgado. Llamó a ese objeto un cuerpo negro perfecto o simplemente "un cuerpo negro". Lo que Kirchhoff mostró es que todos los cuerpos con estas características, con independencia de su naturaleza interna, mantendrían la misma proporción por temperatura y por longitud de onda entre la luz absorvida y la emitida. 

Podríamos anunciar la ley de radiación de Kirchhoff como: 

Para cualquier cuerpo negro en equilibrio térmico a temperatura $T$, la razón entre la potencia absorvida a una cierta longitud de onda $\lambda$ y la potencia emitida en esa misma longitud de onda es $B(\lambda, T)$, con $B$ una función que es independiente de la naturaleza del cuerpo negro.  

La deducción de Kirchoff no es suficientemente formal y es larga, así que la omito aquí, pero podrán consultar el artículo. Lo subiré al bot. 

En su artículo además menciona que sería importante encontrar esa razón entre las potencias de emición y absorción de la luz, tanto experimental como teóricamente, aunque reconoce las dificultades del asunto. 

Aquí debo decir que Kirchhoff no lo sabía (aunque se enteró poco tiempo después, pues en un libro posterior lo menciona), pero un par de años antes que él un físico escosés llamado Stewart ya había hecho algunas mediciones usando la radiación de cuerpos cubiertos con ollín argullendo justamente que estos debían mantener la razón entre la emisión y la absorción para cada longitud de onda, así que servían de referencia. 

No hago la deducción de Kirchhoff aquí, porque es larga, poco formal y como veremos, hay una mejor versión donde de hecho calculamos $B$. Lo importante del asunto es que esto era llamativo. Si se podía medir $B$, se tendría una nueva ley física. 
 

- Ley de (Tyndall)-Stefan-Boltzmann (1864-1879-1884)

$$ u = \sigma T^4$$


<img src="Tyndall.png" alt="" width="400" height="600">



Esto llevó a un físico irlandés llamado John Tyndall a hacer mediciones sobre la radiación del cuerpo negro 4 años después. Él encontró que la energía radiada por un filamento cubierto con ollín (como lo que hacen en el laboratorio de "contemporanea") por unidad de tiempo y por unidad de superficie era proporcional a alguna función creciente de la temperatura. Más tarde Stephan repitió los experimentos de Tyndall y dedujo que esa función era una cuarta potencia. 

La ley se llama de Stefan-Boltzmann porque una década después, el alumno de doctorado prodigio de Stefan, Ludwig Boltzmann hizo una deducción teórica de la fórmula (fue su tesis doctoral). Es curioso, porque Boltzmann pudo haber obtienido la función $B$ usando su distribución, pero se fue por otro camino. Mezcló las ecuaciones de Maxwell con las de termodinámica que ya se tenían en la época. Maxwell mismo para entonces ya había encontrado que en el vacío, por cumplir con sus ecuaciones se formaba una onda cuya velocidad era la velocidad de la luz y que tenían una dirección que apuntaba por el vector de Poynting. Además había encontrado que la fuerza por unidad de área de ese vector era:  

Presión por el vector de Poynting: 

$P = \frac{1}{3} u$


<img src="radiometro.png" alt="" width="400" height="600">


Donde $u$ es la densidad de energía de la onda. Por otro lado:

$$dU = T dS - PdV \Rightarrow \frac{\partial U}{\partial V} + P = T \frac{\partial S}{\partial V} = T \frac{\partial P}{\partial T}$$

Primera ley, lo pasamos a derivadas partciales, y finalmente usamos una relación de Maxwell para pasar de la parcial de S respecto de V a la parcial de P respecto de T. $U = Vu$, así que su parcial con respecto de $V$ es simplemente $u$. Así que esta ecuación se vuelve: 

$$4 u = T\frac{\partial u}{\partial T}$$

Cuya solución es justamente la ecuación de Stephan-Boltzmann, con una constante indeterminada en la ecuación, pero que se encuentra experimentalmente. Por supuesto esto genera en automático la pregunta ¿por qué experimentalmente sale la constante que sale y no cualquier otra? 

Por otro lado, las ideas de Kirchhoff de una única función sobre la radiación emitida por un cuerpo negro dependiendo de la longitud de onda y de la temperatura resonaban en la comunidad de astrónomos, pues ya habían medido desde hacía tiempo el espectro electromagnético del sol y en 1872 se midió el primer espectro electromagnético de una estrella distinta del sol. Se sospechaba que las estrellas se comportarían como cuerpos negros y tener $B$ podía ser de mucha ayuda. Las líneas de absorsión ya se usaban para determinar los materiales de los que estaba compuesta la estrella, tener $B$ iba a ayudar a obtener la temperatura de las estrellas. Aquí apareció Wien. 

- Wilhelm Carl Werner Otto Fritz Franz Wien (premio Nobel 1911)

Wien propuso una solución en 1893 al problema del cuerpo negro y por ella se llevó el premio Nobel de física en 1911. Por poco y no sólo resuelve el problema del cuerpo negro, sino también encuentra la cuantización de la energía. Lo tuvo enfrente, pero no lo vió. Lo que Wien encontró fue: 

$$\rho(\nu, T) = \nu^3 f(\nu/T)$$

Wien hizo 2 deducciones de esta ley. La primera (la de 1893) la hizo usando pura termodinámica. Es larga y obtiene una aproximación de $f$ cuchareando un poco. Quizá se las dejo de tarea guiándolos, por supuesto. Las ideas se parecían en el fondo mucho a las de Boltzmann para deducir la ley de Stefan-Boltzmann. 

Después de ese trabajo, Lummer y Wien hicieron los primeros experimentos para medir la radiación del cuerpo negro y verificar si la ley de Wien se cumplía.

En fin, hago aquí una versión "moderna" de la deducción de Wien. Aclaro que es moderna, porque Wien en realidad tuvo varios pasos que eran poco formales y se basó en la ley de Stefan-Boltzmann para concluir su cálculo. Nosotros seguiremos un camino más formal. Probablemente fue el no seguir la formalidad lo que no le dejó ver que tenía enfrente la cuántica. Aquí va. Primero, pensemos en un gas de fotones. La luz ejerce una presión, así que de alguna forma se antoja pensarla como partículas incluso cuando la vemos como una onda. Ahora, recordemos que en esa época no se había pensado en las distribuciones cuánticas, sino sólo en la versión clásica de Maxwell-Boltzmann. Es decir, el número de ocupación: 

Distribución de Maxwell-Boltzmann del número de ocupación:

$$ \langle n_l \rangle = \exp(-\beta (\epsilon_l - \mu))$$

Pero como los fotones no interactuan entre sí, agregar o quitar fotones no cuesta energía excepto por la energía del mismo fotón, por lo tanto el potencial químico es 0. 

Ahora vamos a calcular la densidad de estados de los fotones. El problema es casi idéntico al del gas ideal usando la ecuación de onda de Schoedinger, pero con algunos detalles. Consideramos la ecuación de onda $N$-dimensional. 

$$\nabla^2 \phi(\vec{r}) + k^2 \phi(\vec{r}) = 0$$

Aquí $\vec{k}$ es el vector de onda y su magnitud se relaciona con la longitud de onda como: 

$$k = \frac{2\pi}{\lambda}$$

Además, la velocidad de fase se relaciona con la frecuencia y con la longitud de onda con: 

Velocidad de fase de la onda:

$$ v = \lambda \nu = \frac{\omega}{k}$$

Recuerden que esta no es la buena velocidad, de hecho esta puede llegar a ser mayor a la velocidad de la luz. La buena velocidad es la velocidad de grupo, que es: 

Velocida de grupo de la onda: 

$$v_g = \frac{d\omega}{dk}$$

Comparar ambas velocidades es lo que se conoce como la relación de dispersión y muchas veces es lo que tendremos experimentalmente. 

Igual que con la ecuación de Schoedinger, llegamos a que la solución de la ecuación de onda es: 

$$\phi(\vec{r}) = A \sin(\frac{n_z \pi z}{L}) \sin(\frac{n_z \pi z}{L}) \sin(\frac{n_z \pi z}{L})$$

con las $n_i$ números enteros positivos, lo que deja: 

$$\vec{k} = (\frac{\pi}{L}n_x, \frac{\pi}{L}n_y ,\frac{\pi}{L}n_z)$$

O  sea, todo sale igualito. Lo repetí sobre todo por la parte de la relación de disperción. De aquí otra vez contamos cuantos estados hay en términos de $k$ desde $k = 0 hasta algún valor dado. Tendremos que el volumen del octavo de esfera es: 

Octavo de volumen de la esfera: 

$$vol = \frac{V k^3}{6\pi^2}$$

Por lo que la densidad de estados en términos de $k$ es: 

$$f(k)dk = \frac{Vk^2dk}{2\pi^2}$$

Cambiando al espacio de $\omega$ da:

$$f(\omega) d\omega = \frac{V k(\omega)^2}{2\pi^2}\frac{dk}{d\omega}d\omega = \frac{V\omega^2 d\omega}{2\pi^2 v^2 v_g}$$

Típicamente $v = v_g$ si es no dispersivo el medio. En otros casos, pues hay que sustituir la relación de dispersión adecuada. 

En el caso de los fotones, tanto $v$ como $v_g$ son iguales a $c$, la velocidad de la luz, por lo tanto: 

$$f(\omega)d\omega = \frac{V \omega^2 d\omega}{2\pi^2 c^3}$$

Aquí vale la pena explicar de nuevo qué es $f(\omega)$ y que es $\langle n_\omega \rangle$. 

Para cada energía, es decir, para cada frecuencia (angular) $\omega$ de un fotón hay $f(\omega)$ posibles estados y cada uno de esos posibles estados podrá o no estar ocupado por un fotón con probabilidad $\langle n_\omega \rangle$. Por lo tanto, la densidad de fotones totales será: 

$$dN_\omega = f(\omega)\langle n_\omega \rangle d\omega = \exp(-\beta \epsilon_l ) \frac{V \omega^2 d\omega}{2\pi^2 c^3}$$

Aquí me estoy comiendo una parte, que es que hay una degeneración del fotón. Hay que multiplicarlo por 2, lo que desparece el "2" en la fracción. En otra clase platico un poco más de esto. 

Ahora, para obtener la densidad energética, sólo se tiene que multiplicar esta densidad de número por la energía de un fotón y aquí es donde Wien en el fondo habla de cuantización, aunque trata de darle la vuelta. Dice que la energía de cada fotón, según la teoría crepuscular de la época, debería de ser proporcional a $\omega$, por lo que mete una constante $a$ que no sabe explicar bien. 

$$\epsilon_\omega = a\exp(-\beta a\omega) \frac{V \omega^3 d\omega}{2\pi^2 c^3}$$

Que pueden sustituir $\omega$ por $2\pi \nu$ y tendrán la relación de Wien de arriba.  

En este punto debo confesar que Wien no hizo esta deducción, pero casi. La densidad de estados no la calculó. obtuvo la integral de $\epsilon$ sobre todas las $\omega$s y esa integral la igualó a $\sigma T^4$. Luego, expandió la integral (junto con la $f$ en series de Taylor con ciertos coeficients desconocidos, que igualando a $\sigma T^4$ se veía que sólo el coeficiente donde hubiera un $T^4$ sobreviviría y con eso obtuvo una approximación a la $f$ salvo una constante. Que es basicamente esa constante $a$ que estoy poniendo. Y de hecho Wien trata de argumentar que sobre la naturaleza crepuscular de los fotones, pero no da el salto completo. Lo tenía enfrente, pero no se atrevió a decirlo o no lo vio.

Aquí entró en el tema otro premio Nobel. 

- Radiación de Rayleigh (premio Nobel 1904)

Rayleigh recibió el premio Nobel antes que los demás, porque su contribución no fue sólo en esto del cuerpo negro, sino más en general sobre la radiación electromagnética, que entre otras cosas explica el cielo azul. Ya estaba trabajando el tema desde hacía unos años, cuando se topó con el conflicto de la teoría de Wien, se le ocurrió inmediatamente que el problema debía venir por estar pensando en las ondas electromagnéticas como crepusculos. Eran ondas, él lo había estado usando ya en todo. Y usar la versión de ondas iba a resolver el problema. 

Rayleigh sí calculó la densidad de estados como lo hicimos nosotros, es decir: 

$$f(\omega)d\omega = \frac{V \omega^2 d\omega}{2\pi^2 c^3}$$

Pero no creyó en el camino crepuscular, así que tampoco pensaba en un número de partículas por estado. Para Rayleigh, eso ya era $dN$ y sólo hacía falta multiplicarlo por una energía. Esa energía, que era la energía por onda, debía ser simplemente $kT$, como en el gas ideal. Equipartición de la energía!! 

$$\epsilon_\omega d\omega = kT  \frac{V \omega^2 d\omega}{2\pi^2 c^3}$$

```julia
ϵ(kT, ω) = ω^3/(2π^2*(exp(ω/kT)-1))
ϵ_w(kT,ω) = exp(-ω/kT)*ω^3/(2π^2)
ϵ_r(kT,ω) = kT * ω^2/(2π^2) 
ωs = 0:0.001:1
i = 0
plot(xlabel = L"$\nu$", ylabel = L"\epsilon")
kT = 0.1
    i += 1
    plot!(ωs, ϵ.(kT, ωs), label = "datos reales", color = i)
    
    plot!(ωs, ϵ_w.(kT, ωs), label = "Wien", color = i, ls = :dash)
    plot!(ωs, ϵ_r.(kT, ωs), label = "Rayleigh", color = i, ls = :dashdot)

plot!(ylim = (0, 8e-5))
```

Listo, problema resuelto! Esta ecuación pegaba perfecto con los datos de Lummer para bajas frecuencias. El problema es que no pegaba para altas frecuencias. 

Así que para frecuencias muy bajas, Rayleigh, para frecuencias altas, Wien. En medio, nadie. Era como que la luz a frecuencias bajas se comportaran como ondas, frecuencias altas como partículas. ¿Qué diantres? Y nadie lograba dar una fórmula para todo el comportamiento. 

Aquí entró Plank al rescate. Plank, un joven que no entendía el problema se le ocurrió una solución fácil. 

- Cuchareo de Plank (Premio Nobel 1918)

Para Plank estaba claro, tenía 2 funciones que pegaban en los extremos, podía promediarlas, pero no exactamente. Promediarlas haría que ganara la que creciera más en cada frecuencia y eso habría sido siempre la versión de Rayleigh. Lo que se quería era que ganar en bajas frecuencias la de Rayleigh y en altas frecuencias la de Wien. Es decir, queríamos que donde la derivada fuera cercana a 0, fuera lo que dominara. Una forma sencilla de hacer eso es aproximar el inverso de la derivada por la suma de los inversos. Así, lo que esté cercano a "0" dominará. Esto lo podía hacer variando la frecuencia o variando la temperatura, pues son las variables del sistema. Planck le apostó a la temperatura. ¿Por qué? quien sabe!! Es decir, hizo lo siguiente:  

$$U_{\omega 1}  =   \frac{1}{ \beta }$$

$$U_{\omega 2} = \exp(-\beta a\omega) a \omega$$

Aquí había un montón de constantes (si lo que vamos a variar es la temperatura), pero son las mismas en ambas ecuaciones (dejo lo que no es igual). Así que esas las sacamos de la interpolación para no cargarlas. Por eso le llamé $U$ y no $\epsilon$

Luego derivó ambas con respecto a $\beta$. 

$$ \frac{\partial U_1}{\partial\beta} = - \frac{1}{\beta^2} = -U_1^2$$

$$ \frac{\partial U_2}{\partial\beta} = - a\omega \exp(-\beta a\omega) a\omega = -a\omega U_2$$



Hago el promedio de estas dos cosas, saco su inverso y con eso aproximo la derivada de:

$$ \frac{\partial \beta}{\partial U} = -\frac{1}{a\omega U + U^2}$$

Finalmente, integro para obtener $U$ en términos de $\beta$: 

$$ \beta = \frac{\log(U)-\log(a\omega + U)}{a\omega}$$

o bien: 

$$U = \frac{a\omega}{\exp(-\beta a\omega) -1}$$

o bien: 

$$ \epsilon_\omega = \frac{a\omega^3}{\exp(-\beta a\omega) -1} \frac{V  d\omega}{2\pi^2 c^3}$$

En realidad no hizo exactamente esto, sino que se sacó de la manga una entropía sin explicar de dónde salía esa entropía y con ella hizo interpolación forzando a que cumpliera con las ecuaciones de Rayleigh y de Wien, pero fue algo similar, sólo más cuchareado aún. Lo curioso es que 2 semanas después salieron un parr de arículos comprobando la fórmula de Plank experimentalmente. Creo que es probable que de hecho el orden de las cosas fuera un poco al revés. Plank conocía las mediciones y cuchareó hasta que le salió una fórmula adecuada. Nunca lo sabremos. Unos meses después justificó vía el microcanónico lo de la entropía que se había sacado de la manga. El microcanónico era tal cual el de las partículas idénticas en cajas idénticas. No explicó porque eran todas idénticas, aunque ahí fue donde habló de la cuantización de la energía, pues no tenía otra forma de explicar el número de cajas. 

Plank siempre justificó ese microcanónico como un artilujio para aproximar la entropía. Nunca creyó que esa versión discreta fuera la correcta, era simplemente la forma que podía calcular y cargaba con una constante que no sabía explicar. Aún así recibió el premio Nobel por esa propuesta de la cuantización de la energía. Un poco una casualidad!!

- Catástrofe ultravioleta

Aunque Plank encontró la fórmula que explicaba todo el comportamiento de la radiación del cuerpo negro en 1901, nadie creía que estaba bien explicada. No se resolvía eso de onda o partícula y no se entendía porque el cuchareo de Plank había pegado. Igual, ya estaba empujándose las ideas cuánticas por todos lados, aunque las grandes celebridades de la época y quienes lo habían desarrollado inicialmente, no creían en esas ideas. Einstein usó la fórmula de Plank para obtener un promedio de ocupación bosónica para calcular el efecto fotoeléctrico, por ejemplo, sin creer en que la fórmula fuera correcta, sólo funcionaba. Así que para 1911 Erenfest, el de las transiciones de fase, llamó a todo este embroyo, una catástrofe ultravioleta, porque tomar a la luz como una onda fallaba al obtener la radiación del cuerpo negro para frecuencias altas y para Erenfest, la luz era una onda, no una partícula. Si hubiera existido Newton, quizá hubiera dicho que se trataba de una catástrofe infraroja. 

El punto es que, aunque ya se tenía la fórmula, aún no se tenía una explicación de esa fórmula. Eso tardó más de 20 años en llegar y llegó, como muchas cosas en ciencia, por accidente. 

- Estadística de Bose y la explicación de la radiación del cuerpo negro (no premio Nobel, aunque 30 nominaciones)


<img src="Bose.png" alt="" width="400" height="600">

Satyendra Bose era profesor en la universidad de Dhaka en la India y había estado leyendo los artículos sobre física estadística y quería llevarles a sus alumnos los cálculos de los grandes físicos de la época. Lo que Bose quería presentar era la catástrofe ultravioleta. Por un lado tenía la versión de Rayleigh y por otra quería hacer la versión de Wien calculando el $\langle n_l \rangle$. Bose intentó hacer el cálculo vía el microcanónico al parecer en un método bastante similar al que hicimos la clase pasada, donde iba a tener $M$ cajas distinguibles y $N$ bolitas que acomodar ahí dentro. Pero se confundió e hizo el cálculo de acomodar $N$ bolitas en $M$ cajas indistinguibles como lo hicimos, así que no le salió la fórmula de Wien, sino la de Plank. 

Según el mismo contó, se fue a su casa pensando en ese error y a revisar los cálculos una y otra vez. Encontró el error y empezó a considerar que no era un error. Para entonces la cuántica aunque aún no estaba fundamentada del todo, ya había demostrado en varios aspectos ser correcta. Escribió un artículo y lo envió a Elsevier para ser publicado. Le rechazaron casi de inmediato el artículo por "no ser interesante". Lo envió a varias revistas inglesas y todas lo rechazaron por no aportar nada. Entonces le escribió a Einstein. 

Como recordarán, Einstein fue alemán y por lo tanto no hablaba inglés perfecto. Así que en algún momento Einstein y Bose se conocieron, cuando Einstein aún no era famoso y Einstein le pidió ayuda a Bose para revisar el inglés de uno de sus artículos. Bose le ayudó y así se hicieron amigos. Esta vez Bose le pidió ayuda a Einstein para traducir el artículo de inglés a Alemán. Bose hablaba algo de alemán, pero no suficiente y quería intentar enviarlo a Springer. Cuando Einstein leyó el artículo le dijo a Bose que no sólo le ayudaría a traducirlo, sino que el mismo lo enviaría a su nombre para recomendarlo. Eso ocurrió en los años veintes, cuando Einstein ya era bastante famoso, así que en la editorial recibieron el artículo de Bose y lo publicaron casi ipso facto. 

El artículo fue escrito 100% por Bose y como único autor. Así que la estadística no es Estadística de Bose-Einstein, sino estadística de Bose, que fue el que dio la idea original. Sin embargo, la historia no acaba ahí. Después de esa publicación Einstein invitó a Bose a escribir un artículo juntos sobre qué sucedía cuando se bajaba la temperatura de un gas de Bosones. En el artículo ambos encontraron el condensado de Bose-Einstein. Así que el condensado sí es de Bose-Einstein. 

Ahora vamos a hacer una deducción alternativa, más de libro y la que al parecer el mismo Bose propuso más adelante en sus libros (no me consta, no los he leído). 

La fórmula de Planck que Bose dedujo constituye, en mi opinión, un avance importante. El método empleado aquí también da lugar a la teoría cuántica del gas ideal, como explicaré en otro lugar.

## Gas de fotones

Vamos a comenzar por obtener el ensamble canónico no vía esos trucos de hacer el ensamble microcanónico y luego sustituir la $\lambda$, que no tenemos certeza de que funcionen excepto en el caso de la aproximación clásica de Boltzmann. Los demás casos quizá no funcionan, así que tenemos que hacerlo directamente.  

$$Z(T, V, N) = \sum_{n_1, n_2, \dots} \exp(-\beta \sum_r n_r \epsilon_r)$$

Aquí los $n_1, n_2, \dots$ satisfacen que: 

$$N = \sum_r n_r$$

Por culpa de esta restricción, la suma de la $Z$ es difícil de evaluar, así que uno tiene que buscar alternativas. No sé si haya una formulita que ayude a resolverla. Hasta ahora no me la he topado ni se me ha occurrido (aunque tampoco la he pesnado demasiado). Se ve difícil. Así que la alternativa es no evaluar esa suma, sino pasarse al ensamble gran canónico!!

$$ \mathscr{Z}(T,V,\mu) = \sum_{N= 0}^\infty Z(T,V,N) e^{\mu N}$$

De primera instancia parece que es más difícil a las $N$ sumas le agregamos una más. Pero esta suma extra hará que todo sea fácil. 

Ahora tenemos que todas las $n_r$'s deben de sumar 0 o deben de sumar 1 o deben de sumar 2, o etc. Esto permite a las $n_r$ ser cualquier cantidad. Entonces: 

$$ \mathscr{Z}(T,V,\mu) = \sum_{N= 0}^\infty \sum_{n_1, n_2, \dots}  e^{\mu (\sum_r n_r)}\prod_r \exp(-\beta n_r \epsilon_r)$$

$$= \prod_r (\sum_{n_r = 0}^\infty  \exp(-\beta n_r \epsilon_r) e^{\mu n_r}) = \prod_r (\sum_{n_r = 0}^\infty   \exp(-\beta n_r \epsilon_r+\mu n_r))  =  \prod_r (\sum_{n_r = 0}^\infty   \exp(-\beta  \epsilon_r+\mu)^{n_r}) $$

Donde la suma es una suma telescópica, porque la exponencial es menor que 1. Por lo tanto: 

Bose:

$$ \mathscr{Z}(T,V,\mu) = \prod_r \frac{1}{1-\exp(-\beta  \epsilon_r+\mu)}$$

Si uno hace el caso fermiónico, entonces las $n_r$'s sólo pueden tener o 0 o 1 como posible valor, y por lo tanto:

FD:

$$ \mathscr{Z}(T,V,\mu) = \prod_r (1+\exp(-\beta  \epsilon_r+\mu))$$


Ahora bien: 

$$\frac{\partial \log(\mathscr{Z}(T,V,\mu))}{\partial \mu} = -\langle N \rangle$$

Que sale igualito que como encontramos que la derivada del $\log(Z)$ con respecto de $\beta$ es la energía promedio o con respecto de $V$ es la presión. 

El logaritmo es la suma de los logaritmos. Cuando derivamos queda: 

Bose (-) FD (+)
$$\sum_r \frac{1}{\exp(-\mu +\beta \epsilon_r) \mp 1} = \langle N \rangle$$

De aquí vemos que para cada energía hay un número asociado. Este número es una proporción del total de partículas (promedio), de donde reconocemos que corresponde cada uno de estos números al número de ocupación: 

$$\langle n_r \rangle = \frac{1}{\exp(-\mu + \beta \epsilon_r)\mp 1}$$ 

Que es lo que ya habíamos calculado con la versión microcanónica y de forma (según yo), más elegante. 

Ya que tenemos esto, encontrar la densidad de energía sigue exactamente los mismos pasos que para la versión de Wien, pero poniendo la distribución de Bose en vez de la distribución de Boltzmann. 

$$f(\omega)d\omega = \frac{V \omega^2 d\omega}{2\pi^2 c^3}$$

$$dN_\omega = f(\omega)\langle n_\omega \rangle d\omega = \frac{1}{\exp(\beta \epsilon_r) - 1} \frac{V \omega^2 d\omega}{2\pi^2 c^3}$$



$$ \epsilon_r = \hbar \omega_r$$

$$\epsilon_\omega = \epsilon_r dN_\omega = \frac{1}{\exp( \beta \hbar \omega) - 1} \frac{\hbar V \omega^3 d\omega}{2\pi^2 c^3}$$

Aquí la $\mu$ se fue, porque son fotones. No se irá cuando sean átomos que condensemos. Si integramos esto con respecto de todas las $\omega$s está un tanto más difícil de hacerse, se puede hacer usando la función $\Gamma$ o usando transformadas de Fourier (se los dejaré de tarea), pero al final se llega a:  

$$ \frac{\epsilon}{V} = \sigma T^4$$ 

con $\sigma = \frac{\pi^2 k_B^4}{15 \hbar^3 c^3}$

Para calcular la presión nos conviene pasarnos a la energía libre de Hemholtz vía una transformada de Lagrange: 

$$\beta F = \beta(E- TS) = \log(\mathscr{Z}(T,V,\mu))+ \beta\mu \langle N \rangle = \log(\mathscr{Z}(T,V,\mu)) = -\log(\prod_r (1-\exp(-\beta  \epsilon_r)))$$

Esto se vuelve una suma que aproximamos por una integral usando la densidad de estados:

$$\approx -\int_0^\infty \frac{V\omega^2 d\omega}{\pi^2 c^3} \log( 1-\exp(-\beta  \hbar \omega))$$

Por lo tanto: 

$$P = \frac{\partial F}{\partial V} \approx - k_B T \int_0^\infty \frac{\omega^2 d\omega}{\pi^2 c^3} \log( 1-\exp(-\beta  \hbar \omega))$$

Esta integral, otra vez está latocita. Se puede hacer por partes y queda: 

$$ P \approx \frac{\sigma}{3} T^4 =  \frac{1}{3} u$$ 

donde $u$ es la densidad energética. Esto es lo que se encuentra vía el vector de Pointing. Aquí dedujimos esto mismo con argumentos termodinámicos!! ;-)

Es decir, la luz tiene una energía, esa energía corresponde a una temperatura también y esa temperatura nos indica la presión que ejerserá. Bonito, no?

Hasta aquí queda hoy. Y hoy dedujimos lo de 3 premios Nobel + un no-premio Nobel que debió ser + otras cosas!! :-).
