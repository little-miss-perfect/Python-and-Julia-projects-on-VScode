## Problema de Monty Hall



La última clase se me pasó ponerles un problema muy famoso, el de Monty Hall. 

Monty Hall era un presentador de un progrma de concursos tipo Chabelo. Al final del programa, le daba la opción al ganador del concurso quedarse con su premio o cambiarlo por uno detrás de 3 puertas. Si el jugador aceptaba, tenía que elegir la puerta donde estaría su nuevo premio. Típicamente 2 de ellos eran de broma, uno interesante. 

Entonces, el jugador decía la puerta 1, la 2 o la 3. Ahí Monty Hall destapaba una de las dos puertas, mostrando uno de los dos premios basura y preguntaba ¿Quieres quedarte con tu puerta, o prefieres cambiar? 

La pregunta es ¿Qué conviene hacer? o ¿Da lo mismo?

Habrá los eventos: 

A1 = Auto en la puerta 1

A2 = Auto en la puerta 2

A1 = Auto en la puerta 3


$P(A1) = P(A2) = P(A2) = 1/3$

Supongamos que elegimos la puerta 3 y sin pérdida de generalidad, supongamos que Monty Hall eligió abrir la puerta 1. 

¿cuál es la $P(A3|1)$ = probabilidad de ganar el Auto dado que Monty Hall abrió la puerta 1?

$P(1|A3) =$ probabilidad de que Monty Hall abra la puerta 1 dado que el auto está en la puerta 3. 

$P(1|A2)$

$P(1|A1)$

Aquí está el secreto. Monty Hall sabe dónde está el premio y no va a abrir la puerta con el premio, porque si lo hiciera, uno cambiaría a esa puerta. Tampoco abrirá la puerta del jugador, porque con toda certeza eso o le mostraría que ya tiene el premio, o le mostraría que no lo tiene y que debe de cambiar obligatoriamente. 

Si el Auto está en la puerta 1, la probabilidad de que abra esa puerta es $0$. Si el auto está en la puerta 2, como nosotros elegimos la 3, Monty está obligado a abrir la puerta 1, así que esa probabilidad es $1$. Si en cambio, el auto está en la puerta 2, la probabilidad de que Monty abra la puerta 1 es $1/2$, porque tiene tanto la puerta 1, como la 2 como opción. 

Ahora podemos aplicar Bayes. 

$P(A3|1) = \frac{P(1|A3)P(A3)}{P(1|A3)P(A3) + P(1|A2)P(A2) + P(1|A1)P(A1)} = \frac{1/2 \cdot 1/3}{1/6 + 1/3 + 0} = \frac{1/6}{1/2} = \frac{1}{3}$

$P(A2|1) = \frac{P(1|A2)P(A2)}{P(1|A3)P(A3) + P(1|A2)P(A2) + P(1|A1)P(A1)} = \frac{1/3}{1/6 + 1/3 + 0} = \frac{1/3}{1/2} = \frac{2}{3}$

Es decir, conviene cambiar de puerta una vez que Monty abre la puerta con el premio "basura". 

Ya que uno lo piensa un rato, es lógico, pero de entrada creo que es sorprendente el resultado. 

## Combinatoria

Ahora sí, de regreso a lo de combinatoria. Quiero comenzar la clase recordando las dos reglas que vimos la clase pasada con cuidado. La regla básica de conteo, que dice que 

### Ley básica de conteo

El número de configuraciones totales de $m$ acciones donde en cada acción hay $r_i$ posibilidades es:

$$\Pi_{i= 1}^m r_i$$

Y también vimos las permutaciones, que consisten en arreglos ordenados de $r$ objetos de entre $m$ objetos. Calculamos que el número de permutaciones es: 

### Número de permutaciones

de $r$ objetos distintos ordenados de entre $m$ posibles: 

$$(m)_r = \frac{m!}{(m-r)!}$$

**Problema:** Supongamos que les pidiera a todos que echaran su celular en una bolsa, porque quiero hacer un examen y decidí que quiero ser estricto. Después, cuando se termina su examen, se los regreso y decido sacar aleatoriamente uno por uno los celulares y entregarlos en el mismo orden en el que los recogí. ¿Cuál es la probabilidad de que reciban su celular? 

Aquí prácticamente nos quedamos. 

Pusimos un ejercicio: ¿Cuál es la probabilidad de que consigan su celular si podemos todos los celulares en una bolsa y después los repartimos aleatoriamente?

Si con esto nos referimos al cual es la probabilidad de que alguien específico reciba su celular, podemos verlo de dos formas. 

Una forma es notar que todas las configuraciones de entregar el celular son justamente $r!$, porque son el número de permutaciones de $r$ objetos de entre $r$ objetos, así que queda $r!/(r-r)! = r!/0! = r!$, pues $0! = 1$. Entre todas esas posibles configuraciones, sólo nos interesa donde la persona específica recibe su celular, y los otros $r-1$ personas no nos importan. De esos casos hay $(r-1)!$, así que la probabilidad de que una persona específica reciba su celular es $(r-1)!/r! = 1/r$.

Otra forma de verlo es pensar que nuestro celular puede quedar en manos de cada una de las personas $r$, de entre esas, sólo en un caso le toca a la persona específica, así que la probabilidad de que eso suceda es $1/r$. Que es la forma fácil de hacer el cálculo. Una forma aún más difícil sería calcular la probabilidad de recibir el celular dado que fui la primera persona en sacar un celular, dado que fui la segunda, etc... Si soy la primera persona, entonces la probabilidad de sacar el celular es $1/r$, si soy la segunda, es la probabilidad de que no lo haya sacado la primera persona $(r-1)/r$ por la probabilidad de que yo lo sacque $1/(r-1)$, que haciendo el producto da $1/r$. Algo similar para el caso donde soy la tercera persona, la cuarta, etc... Todos da $1/r$. En todos los casos lo tengo que multiplicar por la probabilidad de ser la $n$-esima persona, que es también $1/r$. Por lo tanto, la probabilidad de sacar mi celular es $\Sigma_i 1/r^2 = r/r^2 = 1/r$.

Pero también podemos referirnos con esa pregunta a: ¿Cuál es la probabilidad de que todo el grupo obtenga su celular? Para eso notamos que hay $r!$ permutaciones posibles y sólo en una todos obtienen su celular, así que la probabilidad de que todos obtengan su celular es $1/r!$. Si consideramos los que quedan en el grupo, que son 21 personas, esto es más o menos $2\times 10^{-20}$. 

Son sólo 21 personas y ya la probabilidad de que ese evento ocurra es extremadamente raro!! 

Finalmente, aunque ya no es una interpretación posible, hay la probabilidad de que al menos una persona reciba su celular. El problema no es NADA sencillo, así que se los dejo como un reto. Si no pueden calcular la probabilidad con una formula finita, al menos aproximen con una serie. 

**Reto:** ¿Cuál es la probabilidad de que al menos uno de los estudiantes reciba su celular?


En serio inténtenlo... Es un problema retador. 

Ahora sí, llegamos a donde nos quedamos la última clase, las combinaciones. Las combinaciones son como las permutaciones, pero donde no nos importa el orden en el que seleccionamos los $r$ objetos distintos.

### Combinaciones: 

Una combinacioón de $r$ objetos de entre una coleción de $m$ objetos es un arreglo no ordenado de $r$ objetos distintos de entre los $m$ posibles. 

Para calcular el número de combinaciones, notamos que por cada combinación de $r$ objetos, es decir, por cada selección de $r$ de los $m$ objetos, podemos construir $r!$ permutaciones, pues si se tienen $r$ objetos distintos, por la regla para las permutaciones, habrá $r!$ permutaciones. Es decir, el número de combinaciones por $r!$ es el numero de permutaciones, o lo que es lo mismo, el número de combiaciones es el número de permutaciones entre $r!$. 

Hay

$$\binom{m}{r} = \frac{m!}{r!(m-r)!}$$

posibles combinaciones. 

Por ejemplo, si les digo que voy a elegir a 5 personas al azar de entre los que hoy vinieron (para asignarles una labor como equipo), ¿Cuántas posibilidades tengo? 

**5 personas al azar de entre los que hoy vinieron, ¿Cuántas posibilidades tengo?**

Como no me importa el orden, habrá $\binom{N}{5}$, donde $N$ es el número de personas que vinieron hoy. Si son 15, hay unas 3003 posibilidades.

**Demuestra que:**

$P(\cup_{i=1}^N A_i) = \sum_{i=1}^N P(A_i) - \sum_{i<j} P(A_i \cap A_j) +\cdots + (-1)^{n+1} \sum_{k1<k_2<...<k_n} P(\cap A_{k_i}) +\cdots +(-1)^{N+1} P(\cap_{i=1}^N A_i),$

Primero, quiero que notemos que la parte de combinatoria está en las sumas. La primera suma es una suma desde 1 hasta $N$, pero después pusimos los símbolos de menor que. Es decir, en realidad estamos haciendo varias sumas en cada simbolito. ¿Cuantas? Pues un número de combinaciones. Así que cuando la probabilidad de la intersección es una constante para cada configuración de intersecciones, esa súma se vuelve esa constante por el número de combinaciones. 

Dicho esto, pasemos a demostrar esto. 


$P(A\cup B) = P(A) + P(B) - P(A\cap B)$


Primero demostremos el caso con dos conjuntos. Si son disjuntos, no hay nada que probar, porque es parte de la definición de probabilidad, pero si no son disjuntos, tenemos que probarlo.

Para demostrarlo

$P(A\cup B) = P(A) + P(A^c \cap B)$, 

Ya que $A^c \cap B$ es disjunto con $A$ y la unión de ambos es simplemente $A\cup B$. 

También: 

$P(B) = P((A\cap B)\cup (A^c \cap B) = p(A\cap B) + P(A^c \cap B)$, 

por la misma razón. Despejando $ P(A^c \cap B)$ y sustituyendo arriba en la ecuación para $P(A\cup B)$ demostramos la regla para dos conjuntos $P(A\cup B)$. 

La generalización a $N$ conjuntos conviene pensarla por inducción. De hecho, si tenemos 3 conjuntos, podemos aplicar la formula para 2 conjuntos iterativamente y obtener: 

$P(A\cup B \cup C) = P(A\cup B) + P(C) - P((A\cup B) \cap C) = P(A) + P(B) + P(C) - P(A\cap B) - P((A\cap C) \cup (B\cap C))$
$= P(A)+P(B)+P(C)-  P(A\cap B) - P(A\cap C) - P(B\cap C) + P(A\cap B\cap C)$

que lleva a la formula para 3 conjunto. Uno rapidamente adivina que se sigue la fórmula de arriba y que esta se puede demostrar por inducción. Háganlo.
