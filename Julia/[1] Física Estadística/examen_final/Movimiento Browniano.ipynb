# Algunas cuentitas con el gas ideal

Decidí ver primero lo de movimiento Browniano, porque lo de nucleación está inspirado en esto de movimiento browniano. Así que para explicar cómo se forma una nueva fase, primero es buena idea explicar esto de movimiento Browniano. No es indispensable, pero conviene. 

Ahora, para que se entienda porque es poderosa la idea del movimiento Browniano, me gustaría motivar el tema haciendo unas cuentitas sobre el gas ideal aplicado a gases reales. Sólo para tener una idea de las escalas. Y de paso, podemos juntar esos cálculos con lo que vimos de la conductividad eléctrica:  

**Ecuación de Drude**


$$ \sigma_c \approx \frac{N}{V} \frac{e^2 \tau}{m_e}$$

Aquí $\sigma_c$ es la conductividad eléctrica. Puse el subíndice c porque usaré $\sigma$ sin subíndice para la sección eficaz de las partículas.  

El volumen está fijado por el tamaño del sistema. El número de partículas, la carga y la masa atómica, ya se han medido de forma casi directa. Sólo queda entonces medir la $\tau$, es decir, el tiempo libre medio entre colisión y colisión. Hacerlo para un sólido puede ser todo un reto, pero estimar esta cantidad para un gas no es tan complicado y por lo tanto podríamos tener la conductividad de un gas completamente ionizado.  

Comencemos por considerar un tamaño para las partículas y no pensar que son partículas puntuales. 

$$ \delta = \Big(\frac{V}{N}\Big)^{1/3} >> r,$$

Aquí $r$ es el radio de interacción efectivo de las partículas y $\delta$ es el espacio que tiene cada partícula. El radio no es $0$, es sólo pequeño con respecto a la distancia media entre partículas. 



Si pensamos que sólo una partícula se mueve, entonces esta colisionará con todas las que se encuentren en su camino, es decir, todas las partículas que se encuentren en un cilindro cuya altura es $|v|\Delta t$ y su base tiene un área de la sección eficaz de la partícula. Si se trata de partículas duras todas iguales, la sección eficáz es $4\pi r^2$, que es el área de un círculo del doble de radio.  

Por lo tanto, el número de colisiones en un tiempo $\Delta t$ sería: 

$$N_{col} = \frac{N}{V} \sigma |v| \Delta t = \frac{4\pi r^2 N |v| \Delta t}{V} $$

Y el tiempo promedio por colisión es: 

$$\tau = \frac{\Delta t}{N_{coll}} =  \frac{V}{4\pi r^2 N |v|}$$ 

Y para la $|v|$ podemos usar la velocidad promedio, que podemos obtener de la distribución de Maxwell-Boltzmann. 

$$P(v) = \frac{4}{\sqrt{\pi}}\Big(\frac{m}{2k_BT}\Big)^{3/2} v^2 \exp(-m v^2 / 2k_BT)$$

Que es el cálculo que ya hicimos un par de veces. El promedio lo obtenemos de integrar esto multiplicado por $v$ desde 0 hasta infinito, de donde se obtiene que:  

$$\langle v \rangle = \sqrt{\frac{8k_BT}{\pi m}}$$

Sustituyendo en la ecuación de arriva tenemos que: 

$$\tau = \frac{\Delta t}{N_{coll}} =  \frac{V}{8 r^2 N} \sqrt{\frac{ m}{2\pi k_BT}}$$ 

Esto está casi bien, pero no del todo, porque para calcular esto supusimos que sólo nuestra partícula se movía. En realidad todas se mueven, lo que implica no usar la velocidad de la partícula, sino la velocidad relativa. Todas las partículas tienen la misma distribución de velocidades y en cada dirección debe ser una gaussiana (como ya argumentamos antes). Por lo tanto, la velocidad relativa es una velocidad cuya distribución es la de una suma (resta) de dos variables aleatorias con distribución gaussiana. Como vimos, esto da una gaussiana también, pero con una varianza que es la suma de las varianzas. En este caso, la varianza sería del doble, lo que implica cambiar el 2T por 4T en la distribución:

$$P(v_r) = \sqrt{\frac{2}{\pi}}\Big(\frac{m}{2k_BT}\Big)^{3/2} v_r^2 \exp(-m v_r^2 / 4k_BT)$$

Lo que lleva a una velocidad relativa promedio: 

$$\langle v_r \rangle = \sqrt{\frac{16k_BT}{\pi m}}$$

y por lo tanto: 

$$\tau = \frac{\Delta t}{N_{coll}} =  \frac{V}{16 r^2 N} \sqrt{\frac{ m}{\pi k_B T}}$$ 

$$ \sigma_c \approx  \frac{e^2}{16 r^2 \sqrt{m \pi k_B T}}   $$

En un gas completamente ionizado, la conductividad no depende de la densidad de partículas. 

Además, podemos calcular el camino libre medio simplemente multiplicando el tiempo promedio de colsión por la velocidad promedio.

$$\lambda = \langle v \rangle \tau = \frac{V}{\sqrt{2} N  \sigma}$$

Con $\sigma$ la sección eficaz. 

Usaremos esto para aproximar el gas con un caminante aleatorio de paso $\lambda$. 

Sustituyendo los valores típicos de masas ($10^{-27}kg$), radios de partículas ($r \approx 10^{-10}m$), temperaturas ($300K$), y densidades ($10^{23}$ partículas por $m^3$) se obtienen:

$$\lambda \approx 10^{-7}m$$

$$\langle v \rangle \approx 10^2 m/s$$

y 

$$ \tau \approx 10^{-8} s$$ 

Es decir, el camino libre medio es 3 ordenes de magnitud el tamaño de las moléculas. Es como si ustedes pudieran correr en cualquier dirección entre 1 y 2 kms sin estrellarse con nada. Así que realmente prácticamente no hay interacciones. No ver nada 2km a la redonda en promedio. Casi de aquí al centro de Coyacán un vacío y eso en 3D. Pero por otro lado, ocurren $10^8$ colisiones por segundo para cada partícula, así que la escala de tiempo también es importante. En nuestra escala de tiempo el sistema interactua muchas veces y por eso se puede lograr un equilibrio térmico que podemos esperar que sea rápido. Pero ¿Qué tan rápido se logra ese equilibrio total? es decir, el equilibrio de presiones y de partículas. Por ejemplo, si destapo un frasco con algún gas. ¿Cuánto tiempo tardará en difundirse en el salón?

# Caminata aleatoria en cualquier dirección

Para responder esto podemos pensar en un modelo simple, donde las moléculas se mueven a la velocidad promedio el tiempo promedio, es decir, cada $\tau$ se mueven un $\lambda$, sufren una colisión y cambian aleatoriamente su dirección. Esto sería como las caminatas aleatorias que vimos en una red cuadrada o cúbica, excepto que no sólo pueden girar $90^o$ o $180^o$, sino cualquier ángulo. Vamos a suponer que no hay preferencia en la dirección hacia la que cambiarán de posición, sólo es importante que cada cambio de dirección es independiente del anteriror.  

```julia
using Plots, LaTeXStrings
function caminata_aleatoria_1_paso(x, λ)
    θ = rand()*2π
    ϕ = rand()*π
    (x[1]+λ*sin(ϕ)*cos(θ), x[2]+λ*sin(ϕ)*sin(θ), x[3]+ λ*cos(ϕ))
end
```

```julia
gr()

x = (0.,0.,0.)
X = [(copy(x[1]), copy(x[2]), copy(x[3]))]
for i in 1:1000
    x = caminata_aleatoria_1_paso(x, 1)
    push!(X, (copy(x[1]), copy(x[2]), copy(x[3])))
    if mod(i,10) == 0
        plot(key = false, aspect_ratio = 1, grid = false)
        plot!(X, show = :ijulia, color = [RGB(1/1000*j,0,1-1/1000*j) for j in 1:i+1], title = "$i")
    end
    if i == 1000
        plotly()
        plot(key = false, aspect_ratio = 1, grid = false)
        plot!(X, show = :ijulia, color = [RGB(1/1000*j,0,1-1/1000*j) for j in 1:i+1])
    end 
end
```

En este caso, la posición de cada partícula será: 

$$\mathbf{X} = \sum_{i= 1}^n \vec{x}_i$$

Aquí puse en mayúscula la suma, porque es una variable aleatoria y la puse en negrita, porque es vectorial. Las $x_i$ todas tienen norma $\lambda$. Si la distribución de direcciones posibles es aleatoria, entonce el valor esperado de $\mathbf{X}$ será el origen. Ahora, el desplazamiento cuadrático medio será: 

$$\langle X^2 \rangle = \langle \mathbf{X}\cdot \mathbf{X} \rangle = \Big\langle \sum_{i = 1}^n \sum_{j = 1}^n \vec{x}_i \cdot \vec{x}_j \Big\rangle$$

$$= \Big\langle \sum_{i= 1}^n \sum_{j = 1}^n \lambda^2 \cos(\theta_{ij}) \Big\rangle$$

Aquí $\theta_{ij}$ es el ángulo que forman las direcciones entre los vectores $\vec{x}_i$ y $\vec{x}_j$. Si $i = j$ entonces el ángulo es $0$, por lo que podemos re-escribir esto como: 

$$ = \lambda^2 \Big\langle \sum_{i = 1}^n \Big(1+ \sum_{j\ne i} \cos(\theta_{ij})\Big) \Big\rangle = 
\lambda^2\sum_{i = 1}^n 1 + \lambda^2 \Big\langle \sum_{i = 1}^n \sum_{j\ne i} \cos(\theta_{ij}) \Big\rangle $$

La aleatoriedad está en justamente el coseno de $\theta_{ij}$. Pero se trata de una distribución uniforme sobre los ángulos, así que en promedio esa suma debe de dar 0, así que: 

$$\langle X^2 \rangle = n \lambda^2$$

Que es igual al resultado que ya teníamos para la red cuadrada y cúbica. Podemos generalizar aún un poco más esto, permitiendo no sólo pasos de tamaño $\lambda$, sino pasos con alguna distribución, cuyo valor medio es $\lambda$. 

$$ \vec{x}_i = l_i \hat{x}_i$$

Es decir, algún $l$ multiplicado por un vector unitario en la dirección seleccionada. En ese caso: 

$$\langle \vec{x}_i \cdot \vec{x}_j \rangle  = \langle l_i l_j \rangle \delta_{ij} =\langle l^2 \rangle  \delta_{ij}$$

con $\delta_{ij}$ la delta de Kronecker, siguiendo el mismo argumento de omogeneidad sobre los ángulos. Por lo tanto, haciendo el mismo análisis, llegamos a: 

$$\langle X^2 \rangle = n \langle l^2 \rangle$$

La única diferencia es que ahora tenemos el segundo momento de $l$ en vez de $\lambda^2$. Por ejemplo, si tenemos una distribución exponencial:

$$ \rho(l) = \frac{1}{\lambda} \exp(-l/\lambda)$$

Esta es una distribución razonable, pues es pensar que en cada instante tienen una probabilidad independiente de colisionar, así que a lo largo de varios instantes, su probabilidad de colisión será el producto de todas esas probabilidades, es decir $p^n$. Así que el tiempo que puede mantenerse con movimiento libre debe decaer exponencialmente. 

El parámetro de $1/\lambda$ es para que el valor esperado sea justamente $\lambda$. En este caso:

$$\langle l^2 \rangle = \int_0^\infty \frac{l^2}{\lambda} \exp(-l/\lambda) dl = 2\lambda^2$$

Lo que nos da: 

$$\langle X^2 \rangle = 2n \lambda^2 = 2 \lambda (\lambda n) = 2 \lambda \langle v \rangle t$$

Pues $\lambda n$ es la longitud total recorrida en el tiempo $t$, que es $\langle v \rangle t$. De aquí tenemos que el coeficiente de difusión es $\lambda \langle v \rangle/ d$, con $d$ la dimensión del sistema. 

Otra vez, sustituyendo los valores de $\langle v\rangle$ y $\lambda$ tenemos que $D$ es del orden de $10^{-4} m^2/s$.

Para el aire típicamente: 
$$ D \approx 10^{-4} m^2/s$$

Ahora, si queremos saber qué tan rápido se expande/mezcla un gas, tenemos que obtener el radio de difusión, que es la raiz cuadrada del desplazamiento cuadrático medio. En $10^4s \approx 3h$, se expande 1m el gas. En $300h\approx 12 días$ se expandería $10m$. Así que si no hay viento, convección, etc. El proceso es bastante lento. Comparado con el tiempo para que ocurra una colisión, es en extremo lento, podríamos decir, cuasi-estático.

# Partículas Brownianas

Como saben, en 1827, el biólogo Rober Brown vió en el microscopio partículas de Polen y notó que "vibraban". Las observó por largo tiempo y comenzó a dibujar sus trayectorias. Notó entonces que seguían movimientos en zig-zag, de donde primero concluyó que tenían vida y por eso se movían así. Pero más tarde notó que el polvo hacía lo mismo. Eso generó un debate acerca de la razón por la que esto ocurría. 

Debo decir que si intentan ver en el microscopio polen, tienen que ser pacientes. Se mueve más o menos a $10^6 m/s$ cuando es un polen ligero y el agua no está viscosa por otras suciedades. Además, no poner cubreobjetos no deja ver bien, pero poner cubreobjetos frenta un poco el movimiento, así que se vuelve aún más lento. Así que como dije, tienen que ser pacientes en la mayoría de los casos. Es visible el movimiento, pero no es súper claro. Si se trata de polvo, normalmente la masa es mayor y ver el efecto es aún más difícil, así que lo que se observaba es que algunas partículas tenían ese movimiento Browniano, otras no. Explicarlo era el reto y ponía en duda tanto las teorías continuas sobre la materia, como las teorías discretas.

Entonces Einstein llegó, tomó las ideas de un francés llamado Langevin y terminó los cálculos relacionando el movimiento que seguiría con la viscosidad y la temperatura del fluido. 

La idea general es la siguiente. Primero tomemos la Ley de Stokes: 

**Ley de Stokes para una esfera en un fluido:**

$$ F = m \frac{d\vec{v}}{dt} = -6\pi \eta R \vec{v}$$

$R$ es el radio de la esfera, $\eta$ la viscosidad del fluido. La ley dice, la fuerza de fricción es proporcional a la velocidad de la esfera, multiplicado por un coeficiente de fricción que de hecho es proporcional a la viscosidad y al radio de la esfera. 

Esta misma ley es la que se usa para encontrar la velocidad terminal de las cotas de lluvia, por ejemplo o en general la velocidad terminal de un objeto cayendo (igualando esto a $mg$ se tiene que la velocidad límite es $\frac{mg}{6\pi \eta R}$. 

Así que si tenemos una esfera moviéndose, esta se frenará por la fricción. Pero a una escala microscópica, esa misma esfera está recibiendo golpes que la vuelven a acelerar y son de hecho esos golpes los que disipan la energía de la esfera. Esta ecuación nos da una escala de tiempo. La derivada la podemos discretizar y vamos a tener una diferencia en las velocidades dividida entre una diferencia en los tiempos. Por lo tanto, la diferencia en los tiempos va a ir como la diferencia entre las velocidades dividida por lo del otro lado de la ecuación. Es decir: 

$$\Delta t \approx \frac{m \Delta v}{v 6\pi \eta R}$$

$$\Rightarrow t_d = \frac{m}{6\pi \eta R}$$

Aquí pensamos que inicia a una velocidad nula en el tiempo 0. Entonces obtenemos una escala de tiempo donde ya vamos a poder considerar la disipación de la energía. El punto de Einstein fue darse cuenta que esa disipación de la energía es una difusión de la partícula. Se discipa la energía entregándosela a las moléculas, que cuando entran en equilibrio térmico con la partícula Browniana, deben de estar difundiéndola, tal como vimos con el caso del caminante aleatorio continuo (que es un movimiento Browniano). 

Einstein hizo sus cálculos basándose en la ecuación de difusión y una teoría del calor que él mismo había estado impulsando pero es un poco más oscuro que las ideas originales de Langevin y sobre todo que la versión de 1908 de Langevin, donde juntó lo suyo y lo de Einstien para hacer un artículo bastante transparente, así que no voy a hacer lo que Einstein hizo, sino lo que hizo Langevine. 

La idea es: La ecuación de Stokes lleva a que la partícula se detiene, pero eso tiene que ser falso, porque se detiene por el equilibrio termodinámico, que a su vez debe también aportarle energía a la partícula Browniana. Entonces, en vez de tener una rapidez 0 a tiempos largos, la partícula debe de seguir el teorema de equipartición de la energía: 

$$\frac{1}{2}m \langle v^2 \rangle = \frac{k_B T}{2}$$

Entonces, el propone una corrección a la ley de Stokes: 


$$ F = m \frac{d\vec{v}}{dt} = -6\pi \eta R \vec{v} + \mathbf{X}(t)$$

donde $\mathbf{X}$ es una variable aleatoria que tenga valor esperado 0, de forma que no contribuya a la velocidad neta, pero sí a las fluctuaciones de la velocidad. 

Esto por supuesto en ausencia de fuerzas externas. Si ponemos una fuerza externa, entonces hay que sumarla a todo esto. 

## Ecuación de Lagevin

$$  m \frac{d\vec{v}}{dt} = -6\pi \eta R \vec{v} + \mathbf{X}(t) + \vec{F}_e(t, \vec{x})$$

Esta ecuación es súper importante!! La base de las simulaciones de dinámica Browniana.


## Dinámica Browniana

En la parte de la fuerza "externa" uno puede no sólo poner campos externos, sino fuerzas de interacción entre partículas. 

Para ser consistentes con lo que normalmente se maneja, re-escribiré la ecuación de la siguiente forma: 

$$m\ddot{\vec{x}} = - \nabla U(\vec{x}) - \alpha \dot{\vec{x}} + \sqrt{2 \alpha k_\text{B} T} R(t)$$

con 

$\alpha = 6\pi \eta R$


Básicamente cambié la fuerza aleatoria por una fuerza que depende de la temperatura. Lo veremos más adelante (es un paso difícil), pero viene del hecho de que la correlación de $X$ va como $2 \alpha k_B T/m^2$. Así, esto hace que la correlación de $R$ sea 1 y se puede sustituir por una distribución estándar.  

Típicamente en dinámica Browniana se supone que la masa, aunque mucho mayor que la de las moléculas sigue siendo despreciable con respecto del sistema, por lo que lo que se refiere a la aceleración se hace 0. 

$$\frac{d\vec{x}}{dt} = -\frac{1}{\alpha}\nabla U(x) + \sqrt{\frac{2k_B T}{\alpha}}R(t)$$

Que podemos poner en términos del coeficiente de difusión (que también veremos más adelante) como: 

$$\frac{d\vec{x}}{dt} = -\frac{D}{k_B T}\nabla U(x) + \sqrt{2D}R(t)$$

```julia
function euler(f, x0, tf, h = 0.001)
    # resuelve x' = f(t, x)
    xs = [(copy(x0[1]),copy(x0[2]))]
    for t in 0:h:tf
        x0 = x0 + h* f(t, x0)
        push!(xs, (copy(x0[1]),copy(x0[2])))
    end
    return 0:h:tf, xs
end
```

```julia
fuerza(x) = [0.,-2]
f(t,x; fuerza = fuerza, D = 1, kT = 1) = D/kT *fuerza(x)+sqrt(2*D)*randn(2)
ts, xs = euler(f, [0.,0.], 1);
length(xs)
```

```julia
gr()
plot(xs, key = false)
```

```julia
xxs = []
for i in 1:3000
    ts, xs = euler(f, [0.,0.], 0.3);
    push!(xxs, deepcopy(xs))
end

@gif for k in 1:length(xxs[1])
    scatter([xxs[i][k] for i in 1:1000], key = false, color = :black, ms = 0.1)
    plot!(xlim = (-1.1, 1.1), ylim = (-1.,0.1), show = :ijulia, aspect_ratio = 1)
end
```

## Partícula Browniana libre en una dimensión

La ecuación se vuelve: 

$$m\frac{dv}{dt} +\alpha v = X(t)$$


$ \langle X \rangle  = \int_t^{t+\tau} X(t')dt' = 0$ 

Aquí estamos considerando que $\tau$ es mucho mayor que $t_d$, la escala de tiempo donde la ecuación de Stokes se vuelve válda.

Ahora voy a usar la noticación de $\dot{x} = \frac{dx}{dt}$ para simplicar las cosas. Ahora, multiplicamos la ecuación de Langevin por $x$. Entonces: 

$$m \ddot{x}x + \alpha \dot{x}x = x X(t)$$

Ahora, la derivada de $x\dot{x}$ es $x\ddot{x} + \dot{x}^2$.  

$$x\ddot{x} = \frac{d}{dt}(x\dot{x})- \dot{x}^2 $$

$$\Rightarrow m \Big(\frac{d}{dt}(x\dot{x})- \dot{x}^2\Big) + \alpha x\dot{x} = xX(t)$$

Ahora promediamos: 

$$ \langle m\Big(\frac{d}{dt}(x\dot{x})- \dot{x}^2\Big) + \alpha x\dot{x} \rangle =\langle xX(t) \rangle = \langle x \rangle\langle X(t) \rangle = 0$$

Esto porque suponemos que no hay correlación entre la fuerza aleatoria y la posición de la partícula. Sino, no podríamos separar los promedios. Pero en principio no cambia la fuerza aleatoria por dónde se encuentre la partícula, a menos que cambie la temperatura o algo así. Si hubiera gradientes de temperatura, este paso no se podría hacer así nada más. 

Del otro lado, podemos separar el promedio y tenemos: 

$$ m \frac{d}{dt}\langle x\dot{x} \rangle- \frac{1}{2}m k_B T + \alpha \langle x\dot{x}\rangle  = 0$$

Aquí usé el teorema de equipartición para la parte de $k_B T$. Esta ecuación es una ecuación de primer grado fácil de resolverse, así que por formulazo llegamos a: 

$$\langle x\dot{x} \rangle = C\exp(-\gamma t) +\frac{k_B T}{\alpha}$$

$$\gamma = \alpha/m$$

si en $t = 0$ $x = 0$, entonces:

$$\langle x\dot{x} \rangle = \frac{1}{2}\frac{d}{dt}\langle x^2 \rangle = \frac{k_B T}{\alpha}\Big(1- \exp(-\gamma t)\Big)$$

Que otra vez es una ecuación sencilla de resolver. Pues podemos integrar directamente. 

$$ \langle x^2 \rangle = \frac{2k_B T}{\alpha}\Big(t-\frac{1}{\gamma}(1-\exp(-\gamma t))\Big)$$

Por lo tanto, para tiempos largos, esta se vuelve la ecuación de difusión!! con el coeficiente que depende de la viscosidad. 

$$D = \frac{k_B T}{6\pi \eta R}$$

Pero además, tenemos el caso de $t \approx 0$ nos queda: 

$$ \langle x^2 \rangle \approx \frac{2k_B T}{m} t^2$$

Es decir, para tiempos cortos la partícula browniana sigue un movimiento balístico!!

Esto se puede usar para aproximar el número de Avogadro experimentalmente, pues se mide hasta qué tiempo hay este comportamiento balístico y luego se compara con la $\tau$ que calculamos arriba y que depende del número de particulas. 

Ahora regresemos a nuestra ecuación original con las velocidades: 

### Momentos de la velocidad

$$\frac{dv}{dt} +\gamma v = X(t)/m$$

Esto podemos "integrarlo" directamente: 

$$v(t) = v_0 \exp(-\gamma t) + \frac{\exp(-\gamma t)}{m}\int_0^t dt' \exp(-\gamma t) X(t')$$

Si sacamos el promedio, nuevamente se va a 0 lo que está dentro de la integral y nos queda sólo la exponencial: 

$$\langle v(t) \rangle = \langle v_0 \rangle \exp(-\gamma t)$$

O sea, la velocidad promedio decae exponencialmente. Para el cuadrado de las velocidades es un poco más difícil. 

$$\langle v^2(t) \rangle = \langle v_0^2 \rangle \exp(-2\gamma t) + \frac{\exp(-2\gamma t)}{m^2}\int_0^t ds \int_0^t ds' \exp(-\gamma(s+s')) \langle X(s)X(s')\rangle $$

Nuevamente se fue la integral que era sencilla. Haciendo cambio de variable:

$$\langle v^2(t) \rangle = \langle v_0^2 \rangle \exp(-2\gamma t) + \frac{\exp(-2\gamma t)}{m^2}\int_0^t ds \int_s^{t+s} dr \exp(-\gamma r) \langle X(s)X(r-s)\rangle $$

Ahora aproximamos el promedio por uno que sea independiente del tiempo, es decir, por uno equivalente a "en equilibrio". Si es invariante de tiempo, podemos recorrer:

$$ \langle X(s)X(r-s)\rangle =  \langle X(s)X(s+r)\rangle = K(r)$$

Esta es la función de autocorrelación de $X$. Cumple con algunas propiedades: 

$$K(0) = \langle X(s)X(s)\rangle = \langle X(s)^2\rangle > 0$$ 

 $$|K(r)| \le K(0)$$

Esto viene de $\langle (X(t) \pm X(t+r))^2\rangle = 2\langle X(t)^2\rangle \pm 2\langle X(t)X(t+r) \rangle \ge 0$, que implica directamente esta escuación. 

Además, 

si $r\rightarrow \infty$

$$K(r) \approx \langle X(r)\rangle \langle X(s+r)\rangle = 0$$

Esto puesto que si $r$ es muy grande, no hay correlación entre los tiempos. Y finalmente: 

$$ K(r) = K(-r)$$

Que se da por ser en equilibrio, así que a un tiempo $t$ la correlación con un tiempo $t+r$ es lo mismo que $t-r$ y $t$. 

Es decir, la función de correlación es simétrica, tiene un máximo en $0$ y decae a 0 para tiempos grandes. Esto nos permite cambiar la $K(r)$ por $K(w)$ con $w = s'-s$, y por lo tanto separar la integral. Además. 

$$\frac{1}{m^2}\int_0^{2t}  dr \exp(-\gamma r)  \int_{-\infty}^{\infty} dw K(w)  = \frac{C}{\gamma}\Big(\exp(-2\gamma t) - 1\Big)$$

Donde la $C$ es la integral de la función de correlación. Esto nos lleva a:

$$\langle v^2(t) \rangle = \langle v_0^2 \rangle \exp(-2\gamma t) + \frac{C \exp(-2\gamma t)}{\gamma}\Big(\exp(-2\gamma t) - 1\Big)$$

En el límite de $t\rightarrow \infty$ 

si  $t\rightarrow \infty$ 

$$\langle v^2(t) \rangle \sim \frac{k_B T}{m} = \frac{C}{2\gamma}$$


Lo que implica

$$ C = \frac{1}{m^2}\int_{-\infty}^\infty K(w)dw = \frac{2\alpha k_B T}{m^2}$$

Es decir, relacionamos la disipación de la energía producida por la viscosidad y la temperatura con la correlación entre los choques. 

```julia
using Plots, LaTeXStrings

Rx(θ) = [1 0 0;0 sin(θ) -cos(θ); 0 cos(θ) sin(θ)]
Ry(θ) = [sin(θ) 0 -cos(θ);0 1 0; cos(θ) 0 sin(θ)]
Rz(θ) = [sin(θ) -cos(θ) 0; cos(θ) sin(θ) 0; 0 0 1]
R(θ, ϕ, ψ) = Rx(θ)*Ry(ϕ)*Rx(ψ)
function circle(r,x,θ, ϕ, ψ) 
    xs = [R(θ, ϕ, ψ)*[r*cos(θ0), r*sin(θ0), 0] .+ x for θ0 in 0:π/50:2π]
    push!(xs, R(θ, ϕ, ψ)*[r, 0, 0] .+ x)
    xs2 = [(xx[1],xx[2],xx[3]) for xx in xs]
    return xs2
end

function cilinder(r,x,h,θ, ϕ, ψ) 
    RR = R(θ, ϕ, ψ)
    xs = circle(r, x, θ, ϕ, ψ)
    xs2 = circle(r, x+RR*[0,0,h], θ, ϕ, ψ)
    xx = [[xs[i][1],xs2[i][1]] for i in 1:length(xs)]
    yy = [[xs[i][2],xs2[i][2]] for i in 1:length(xs)]
    zz = [[xs[i][3],xs2[i][3]] for i in 1:length(xs)]
    return xs, xs2, xx, yy, zz
end

    
```

```julia
r = 2.1
x = [0,0,0]
h = 10.0
θ, ϕ, ψ = 0.3,0.1,0.
xs, xs2, xx, yy, zz = cilinder(r,x,h,θ, ϕ, ψ) ;
length(xx)
```

```julia
plotly()
color = RGB(1,0,0)
plot(xs, key = false,color = color)
plot!(xs2, key = false,color = color)
for i in 1:length(xx)
    plot!(xx[i], yy[i], zz[i], color = color, alpha = 0.1)
end
 plot!(axis = false, grid = false, ticks = false)
scatter!([(4*rand()-2, 6*rand()-4, 10*rand()-10) for i in 1:30], color = :black)
```
