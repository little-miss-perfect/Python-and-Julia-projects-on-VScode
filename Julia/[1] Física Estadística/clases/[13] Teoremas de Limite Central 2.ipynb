# Historia

Quiero comenzar la clase con un poco sobre la historia de los teoremas de límite central, aprovechando que uno de sus compañero me hizo notar (y por lo que asegura ya 1pto sobre su próximo exámen), que la demostración del teoremal del límite central que les di no es de Gauss, sino una versión intermedia entre lo que hizo Chevichev y lo que hizo Polya. 

Entonces, quiero aclarar algunas cosas. Priero, no es "EL teorema del límite central", sino los teoremas de límite central. Se llaman así bautizados por Polya, quien notó que eran una cosa importante. Teoremas sobre el comportamiento estadístico de una variable al aplicarla repetidamente. El primero de ellos fue también el primero que vimos, el del límite de una bariable binomial. Ese teorema lo demostró Abraham de Moivre, en 1733. Por desgracia pasó algo desapersivido. 



$$\int_{-\infty}^\infty \exp(-t^2) = \sqrt{\pi}$$

Casi 80 años después, Laplace descubrió el trabajo de Moivre y sorprendido por el resultado, lo republicó, pero tampoco tuvo mucho éxito. Sin embargo, en el trabajo de Laplace, no sólo publicó de nuevo lo que había hecho Moivre, sino que agregó algunos cálculos más. Por ejemplo, ahí obtuvo la serie de tan(x) que usó para estimar $\pi$, y también obtuvo .

Más o menos por la misma época, Gauss publicó un trabajo sobre combinatoria y la medición de errores. En su trabajo Gauss estudió los errores en una medición de un experimento. Supuso que cada medición es independiente de las demás, por lo que la probabilidad conjunta de obtener cada uno de los valores que se obtiene en la medición es el producto de las probabilidades. Después, usando que la probabilidad conjunta debería de maximizarse al rededor del promedio de las medicioens, logró dar argumentos buenos sobre que la mejor distribución de los errores debería ser la distribución normal bajo determinadas condiciones.

Esta es otra versión de teorema de límite central. La distribución de errores con respecto del valor esperado de una medición, sigue una distribución gaussiana. 

"La distribución de errores con respecto del valor esperado de una medición, sigue una distribución gaussiana."

El trabajo de Gauss no fue realmente formal y de hecho tiene algunos errores de argumentación, pero fue el que tuvo peso durante los siguientes 80 años. Hacia finales del siglo XIX, Maxwell obtuvo vía el desarrollo de la física estadística, que la velocidad de las partículas de un gas debía seguir una distribución normal, similar a la que planteó Gauss sobre sus errores y poco después Lyapunov se topó con el trabajo de Laplace y conectó la distribución de errores de Gauss, la de velocidades de Maxwell, con el límite de la distribución de una variable binaria, aunque no demostró nada tan especial.

Luego, ya en los 1900's, Lyapunov notó que había una relación entre lo de Maxwell, Gauss Laplace (de Moivre) y comenzó a estudiar el tema. Demostró formalmente lo que había hecho Laplace menos formalmente, pero se le adelantó Polya con una demostración similar a la que vimos nosotros para el caso general de variabels idénticas en los años 20's. ## Corolario: 

Una variable aleatoria binaria tiende a una distribución normal cuando $n$ tiende a infinito

### Prueba:

$$X = \begin{cases}
			0, & \text{if $E$ no ocurre }\\

Ahora aplicamos el teorema del límite central que acabamos de ver:

$$\lim_{n\rightarrow \infty} P\Big((\sum X_i -n\mu)/\sigma \sqrt{n}\le x\Big) = \Phi(x)$$

$\mu = \langle X\rangle  = p+ 0(1-p) = p$ y $\sigma^2 = \langle X^2 \rangle -p^2 = p - p^2 = p(1-p) = pq$ 

$$\lim_{n\rightarrow \infty} P\Big(\sum X_i \le x\sqrt{ \sigma^2 n} + n\mu\Big) = \lim_{n\rightarrow \infty} P\Big(\sum X_i \le x\sqrt{ p(1-p)n} + np\Big) = \Phi(x) = \int_{-\infty}^x \frac{1}{\sqrt{2\pi}}\exp(-t^2 /2)dt$$

Pero la suma de estas variables aleatorias es justamente la variable binomial, que haciendo haciendo el cambio de variable $t = (u-np)/\sqrt{p(1-p)n}$ llegamos a justamente a: 

$$\lim_{n\rightarrow \infty} P\Big(\sum X_i \le y \Big) = \Phi(x) = \int_{-\infty}^y \frac{1}{\sqrt{2\pi np(1-p)}}\exp(-(u-np)^2 /2 np(1-p))du$$

Que era lo que queríamos probar!!

Ahora por favor, que alguien pase a obtener la gaussiana que le corresponde a la suma de variables aleatorias con distribución: 

normal y uniforme

- estándar

- uniforme entre 0 y 1

Para el caso de la distribución normal es extremadamente directo. $\sigma = 1$ $\mu = 0$, así que

$$\lim_{n\rightarrow \infty} P\Big(\sum X_i\le x  \sqrt{n}\Big) = \Phi(x) =  \int_{-\infty}^x \frac{1}{\sqrt{2\pi}}\exp(-t^2 /2)dt$$

Nuevamente, haciendo cambio de variable $u = t\sqrt{n}$ $dt = du/\sqrt{n}$

$$\lim_{n\rightarrow \infty} P\Big(\sum X_i\le y\Big) = \Phi(x) =  \int_{-\infty}^y \frac{1}{\sqrt{2\pi n}}\exp(-u^2 /2n)du$$

que es una gaussiana con $\mu = 0$ y $\sigma^2 = n$. 

Si promediamos, obtenemos una variable con una desviación estándar que se reduce como $1/\sqrt{n}$. 

Para la distribución uniforme $\mu = 1/2$, $\sigma^2 = \int_0^1 x^2 dx -1/4= 1/3-1/4 = 1/12$. Por lo tanto: 

$$\lim_{n\rightarrow \infty} P\Big(\sum X_i\le x\sqrt{n}/\sqrt{12} + 1/2\Big) = \Phi(x) =  \int_{-\infty}^x \frac{1}{\sqrt{2\pi}}\exp(-t^2 /2)dt$$

cambiando variables


$$\lim_{n\rightarrow \infty} P\Big(\sum X_i\le y\Big)  =  \int_{-\infty}^y \frac{\sqrt{12}}{\sqrt{2\pi n}}\exp(-6(u-1/2)^2 /n )du$$

2. 10 números reales se redondean y después se suman. ¿Cuál es la probabilidad de que si primero los sumamos y después redondeamos, obtengamos el mismo número?

Llamemos $X_i$ la diferencia entre el redondeo del i-esimo número y el i-ésimo número. Esta será una variable aleatoria uniforme entre $-0.5$ y $0.5$. Nuestras dos sumas redondeadas serán iguales si y solo si, $|\sum X_i| < 0.5$, queremos determinar la probabilidad de que esto ocurra. 

$\mu = 0$ y $\sigma^2  = \langle x^2 \rangle = \int_{-0.5}^{0.5}x^2 dx =2 \cdot 1/3 \cdot 1/8 = 1/12$

así que 

$$P\Big(\sum X_i\le y\Big)  \approx  \int_{-\infty}^y \frac{\sqrt{12}}{\sqrt{2\pi n}}\exp(-6u^2 /n )du$$


$$P\Big(\sum X_i\le -y\Big)  \approx  \int_{-\infty}^{-y} \frac{\sqrt{6}}{\sqrt{\pi n}}\exp(-6u^2 /n )du$$

$$P\Big(|\sum X_i|\le 0.5\Big)  \approx \int_{-\infty}^{0.5} \frac{\sqrt{6}}{\sqrt{\pi n}}\exp(-6u^2 /n )du - \int_{-\infty}^{-0.5} \frac{\sqrt{6}}{\sqrt{\pi n}}\exp(-6u^2 /n )du = \int_{-0.5}^{0.5} \frac{\sqrt{6}}{\sqrt{\pi n}}\exp(-6u^2 /n )du$$

# Entropía

En la parte de termodinámica discutimos un poco sobre este término, que no es otra cosa que la medida de calor aprovechable para hacer un trabajo, o más bien, del calor no aprovechable. Tenemos una cierta cantidad de energía dentro del calor, pero no todo es aprovechable por alguna extraña e incomprensible razón. Queda claro cuando uno piensa en el calor radiado como luz. Tiene la misma energía muchos fotones con baja frecuencia, que un solo fotón con alta frecuencia. 

Si tomamos un fotón de alta frecuencia y le pegamos a un átomo adecuado, este subirá sus niveles energéticos y al regresar a su estado "base", puede que lo haga emitiendo un fotón con la misma frecuencia, o que lo haga varios fotones de menor frecuencia. La cosa es que para que luego me emita un fotón de alta frecuencia, yo no puedo lanzar fotones de menor frecuencia que sumen la misma energía. Lo puedo hacer, pero es difícil atinarle en tiempo al mismo átomo, lo que hace que sea complicado recuperar mi fotón de alta frecuencia. Obtengo la misma energía de regreso, pero en mayor intensidad y menor frecuencia. Así que de alguna forma, espero poder hacer más cosas con los fotones de alta frecuencia que con los fotones de baja frecuencia, aunque tenga en energía la misma cantidad. 

Y así, microscópicamente hablando, Boltzmann se le ocurrió que la entrópía de hecho medía el número de configuraciones de un sistema. Es un poco un misterio cómo llegó a esa conclusión. Una especie de inspiración divina. Sí da algunos argumentos, pero no mucho mejores que los que acabamos de dar. La idea era: Si tengo una energía dada, puedo encontrarme en el caso donde tengo un solo fotón con una frecuencia altísima, donde sólo tengo una configuración, o si voy bajando la frecuencia que permito, voy abriendo paso a más y más configuraciones posibles. Ahora bien, queremos que la entropía cumpla con ser una función aditiva. El número de configuraciones no es aditivo. Si tengo un sistema con un número de posibles configuraciones $\Omega_1$ y otro sistema con $\Omega_2$ configuraciones, entonces el número de configuraciones de los dos sistemas juntos es $\Omega_1 \Omega_2$ si ambos son independientes. Entonces, ¿Cómo volvemos una cantidad que es multiplicativa aditiva? 

Si 2 sistemas tienen $\Omega_1$ y $\Omega_2$ número de configuraciones posibles, ¿cuántas configuraciones tengo si estos sistemas son independientes? 

$$\Omega = \Omega_1 \Omega_2$$

Es decir, el número de configuraciones es multiplicativo para sistemas independientes. ¿Qué cantidad sería aditiva?

$$S(\Omega) = k_b \log(\Omega) = k_b\log(\Omega_1) + k_b\log(\Omega_2) = S(\Omega_1) + S(\Omega_2)$$ 

Y así, Boltzmann nos dió una fórmula para obtener la entropía de Clausius a partir de información microscópica!! 

Vamos a probarla!! 

Boltzmann no tenía la mecánica cuántica, por lo que "el número de configuraciones" era realmente el volumen del espacio fase. Eso lo hace algo más complicado de entenderse, así que por ahora voy a olvidarme que existe el espacio fase. Luego regreso a ello. Queremos comenzar sencillo y en este caso, sencillo es usando mecánica cuántica y no mecánica clásica. Igual regresaremos a ello, porque intentarlo desde la mecánica clásica arroja un resultado incompleto. Nos da una constante paramétrica que no nos podemos deshacer, pero también nos da un experimento para buscar esa constante y lllegar a la conclusión de que algo raro pasa (y ese algo raro es que el mundo es cuántico). 

En fin, entonces, usemos cuántica. 

### Ejemplo de sistema físico: 

Una partícula en una caja de potencial infinito 1D de largo $L$, masa $m$, puntual, cuántica. 

En realidad no vamos a trabajar con una partícula, sino con $N$ y no va a ser en 1D, sino 3D, pues lo que quiero es que estudiemos el gas ideal. Pero comencemos por lo sencillo. 

$$\psi(x) = A \sin(kx)$$ 

donde  $A$ y $k$ son constantes.

Como el potencial es infinito, en $x = 0$ y $x = L$, la función de onda debe ser $0$. 

$ A \sin(0) = 0$ y $A \sin(kL) = 0$

No queremos que $A = 0$, porque eso sería que no hay partícula, así que $kL = \pi n$ con $n$ en los naturales. 

$ k = \frac{\pi}{L}n \Rightarrow k = \frac{2\pi}{\lambda}$,  $\lambda = \frac{2L}{n}$ 

donde $\lambda$ es la longitud de onda. Ahora, usamos que: 

$p = \hbar k = \frac{h}{\lambda}$

De aquí podemos calcular la energía: 

$$E = \frac{1}{2}\frac{p^2}{m} = \frac{\hbar^2 k^2}{2m} = \frac{\hbar^2 \pi^2 n^2}{2m L^2} = \frac{h^2}{2m\lambda^2} = \frac{n^2 h^2}{8mL^2}$$

Cualquiera de las últimas 3 versiones sirve, dependiendo de lo que querramos hacer. 

En ND: 

$$E = \frac{\pi^2 \hbar^2}{2m}\Big(\sum_{i = 1}^N \frac{n_i^2}{L_i^2}\Big)$$

donde las $n_i$'s y $L_i$'s, son las componentes en el eje $i$ de los vectores $\vec{n}$ y $\vec{L}$. El de $L$ describe las dimensiones de la caja, el otro son sólo números cuánticos en cada dirección. 

No hago la deducción, porque es sólo talachuda, así que se las dejo de tarea. Es la versión vectorial de lo que hice aquí. 

Ahora, en realidad lo que nos interesa no es la energía, sino el número de configuraciones y aquí está el secreto. Pues vamos a fijar la energía y vamos a decir qué valores de las $n$'s podemos tener. 

Como las $n$'s son números naturales sólo para ciertos valores de la energía tengamos configuraciones posibles y en general resolver el problema de qué energías son las buenas es difícil. Para el caso de una elipse, no un círculo, es lo que se usa para demostrar el teorema de fermat. Puenden imaginar que es difícil el tema. Pero no nos importa, somos físicos, así que vamos a aproximar. Es lo que sabemos hacer.  

Despejemos las $n$'s y para simplificarnos más las cosas, que sea una caja cúbica. 

$$\sum n_i ^2 = \frac{2m L^2 E}{\pi^2 \hbar^2} = R^2$$ 

Ahora, en vez de pensar en exactamente la energía $\epsilon$, pensamos en $\epsilon \pm \Delta \epsilon$, y pasamos a la aproximación continua. En ese caso, podemos calcular la cumulativa de configuraciones posibles y después derivarla. 

Dibujar lo que queremos contar

Al final, la cumulativa es proporcional al volumen de la esfera de Radio $R$, que es $\frac{4}{3}\pi R^3$

$$ \phi(\epsilon) = \frac{4\pi}{3}\Big(\frac{2mL^2\epsilon}{\pi^2 \hbar^2}\Big)^{3/2} = \frac{4L^3 (2m\epsilon)^{3/2}}{3\pi^2 \hbar^3} = \frac{L^3 (8m\epsilon)^{3/2}}{6\pi^2 \hbar^3} = \frac{V}{6h^3} \pi (8m\epsilon)^{3/2}$$

Esto es para una partícula. Para $N$ partículas $E = \sum \epsilon_i$. Vamos a asumir que la energía en promedio es $\epsilon$ y que cada partícula tiene su energía cercana al promedio $\epsilon$. La cumulativa en general va a ser el producto de las cumulativas. Ese producto se maximiza cuando la distribución de las energías sigue una gaussina (como veremos más adelante). Esta gaussiana se vuelve muy picuda dado el teorema del límite central, por lo que la energía promedio en esta suma es casi una delta de dirac, por lo que asumir que todas las energías son prácticamente la misma, no es tan descabellado. En realidad no lo son, pero la varianza del promedio decae como $1/\sqrt{N}$ y $N \propto 10^{23}$, así que la varianza es alrededor de $10^{-11}$ en unidades de la energía promedio.

En ese caso:

$$ \Phi(E, N, V) \approx  \phi(\epsilon, 1, V)^N$$

Que aproximadamente es el número de configuraciones con energía menor que $E$. 

y por lo tanto 

$$\Omega(E)dE = \frac{d\Phi}{dE}dE = N\phi^{N-1}\frac{d\phi(\epsilon)}{dE}dE $$

$$  N\phi^{N-1}\frac{d\phi(\epsilon)}{d\epsilon} \frac{\partial \epsilon}{\partial E}dE = \phi^{N-1}\frac{d\phi(\epsilon)}{d\epsilon}dE$$ 

Pues la parcial de $\epsilon$ con respecto de $E$ es $1/N$, ya que $E = N\epsilon$. Sustituyendo, queda: 

$$=  \Big(\frac{4V}{3h^3} \pi (2m\epsilon)^{3/2}\Big)^{N-1}\Big( \frac{2V}{h^3} \pi (2m\epsilon)^{1/2}\Big)dE$$

Por lo que la entropía es: 

$$S = k_b \log(\Omega) = k_bN\log(V)+ k_b\frac{3}{2}(N-1)\log(\epsilon)+k_b\frac{1}{2}\log(\epsilon)+ k_b(N-1)\log(4\pi (2m)^{3/2}/3h^3) + k_b\log(2V\pi\sqrt{2m}/h^3) $$

$$ \approx k_bN\log(V) +k_b\frac{3}{2}N\log(E) + f(N)$$

Ahora obtenemos la temperatura y la presión de la ecuación $TdS$

De la primera ley en su forma de entropía tenemos: 

$TdS = dW + dE$

$\Rightarrow \frac{\partial S}{\partial E} = \frac{1}{T}$

y

$\frac{\partial S}{\partial V} = \frac{P}{T}$

Que también se pueden obtener de las relaciones de Maxwell. Entonces: 

$$\frac{\partial S}{\partial V} = \frac{\partial }{\partial V} \Big(k_bN\log(V) +k_b\frac{3}{2}N\log(E) + f(N)\Big)$$

$$= \frac{k_b N}{V}$$

De aquí que: 

$$PV = N k_b T$$

Esto lleva a la ecuación de estado, porque la dependencia con $V$ no cambia, pero la dependencia con $E$ sí cambia, lo que a su vez lleva a calcular $\partial S/\partial E = \frac{1}{2} Nk_B /E = \frac{1}{T}$, pero esto lleva a $E =  \frac{1}{2} Nk_B T$, que es incorrecto. 

El error está en varios libros, porque es correcto para una sola partícula. Como lo más frecuente es hacer el cálculo no con $E$ constante, sino con $T$ constante, la versión de una partícula sirve (como veremos más adelante al estudiar el ensamble canónico). También se propaga el error, porque no afecta para obtener la ecuación de estado y normalmente no se usa el microcanónico para obtener la energía interna. En fin, tengan cuidado con ese error. 

Espero que esto les sorprenda!! acabamos de deducir el gas ideal a partir de información microscópica sólo de asumir como correcta la definición de la entropía de Boltzmann!! 

Esto de poner el número de contar el número de estados como función de la energía, el número de partículas y el volumen, es lo que se conoce como el ensamble microcanónico. 

Ya que vimos que la entropía de Boltzmann fue exitosa en su primer cálculo, discutamos un poco mas sobre porque está bien. 

En su curso de termodinámica debieron ver que cualquier función que cumpla con 3 características: 

- Ser diferenciable, 
- aditiva, 
- e incrementar al reducir las constricciones

Si cumple con esas 3 característica entonces es una entropía que cumple con la definición de Clausius. Hay un sin fin de ellas. 

Nuestra función, como la estamos aproximando en el límite continuo, es diferenciable. Es aditiva, pues el número de configuraciones es multiplicativo. Sólo falta checar que incrementa al quitar constricciones. 

Supongamos que tenemos un gas separado en 2 volúmenes. Entonces el número de configuraciones será el producto de los dos números de configuraciones, como ya vimos. Al retirar una constricción, por ejemplo, permitiendo que cambie el volumen, habría además de las configuraciones que ya se tenían, todas las de la mezcla, así que la entropía aumenta.
