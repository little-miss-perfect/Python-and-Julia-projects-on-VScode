## Estrategia para resolver un problema con ensamble canónico partículas ideales

La estrategia en general es la siguiente: 

0. Obtenemos la densidad de estados $\rho(E)$ o bien $\Omega(E)$ si vamos a trabajar con algo discreto para una sola partícula. 

Este primer punto lo pongo como 0, porque muchas veces lo tenemos dado o es una de las densidades que ya calculamos. 

Después tenemos que calcular el ensamble:

1. Primero obtenemos el ensamble canónico de 1 partícula, es decir $$Z(T,V,1) = \int_0^\infty \rho(E) e^{-\beta E} dE$$ 

También está la versión discreta, con una suma y en vez de $\rho$ es $\Omega$, que son equivalente. A partir de aquí hay varias alternativas. Les pongo las más comunes. 

2. Identificamos si se trata de partículas indistinguibles o distinguibles. Si se puede etiquetar a las partículas, son distinguibles. 

2.1 Si lo que queremos es calcular promedios (como el de la energía) muchas veces podemos hacerlo vía $p_r = e^{-\beta E_r}/Z$ o bien eso por su degeneración. Una vez que tenemos $p_r$ o su versión continua $\rho(T)$, podemos calcular lo que querramos. Así que valoramos si queremos tomar esa vía. Si es así, podemos por ejemplo calcular la energía promedio vía $\langle E \rangle = \sum p_r E_r$. De ahí pasar al paso 4.1.

3. Obtenemos $Z(T,V,N) = Z(T,V,1)^N$ si son partículas distingibles, o $Z(T,V,N) = \frac{1}{N!} Z(T,V,1)^N$ si son partículas indistinguibles.  

4. Obtenemos $F(T, V, N) = k_B T \log(Z(T,V,N)$ y con sus derivadas obtenemos todo lo que necesitemos (vía las relaciones de Maxwell o las relaciones de la ecuación $TdS$. 

4.1. Calculamos $C_v$, susceptivilidad, etc. Vía las derivadas del promedio que hayamos hecho. 

### Sólido de Einstein. 

Como primer ejemplo vamos a estudiar un modelo muy sencillo de cristal. Vamos a pensar en los cristales como lattices donde los vértices son los átomos y estos no interactuan con nadie (es un sistema ideal). Los átomos están "atados" a los vértices de la lattice mediante resortes, de forma que están oscilando como osciladores armónicos. Podemos tratar este problema como que cada átomo es un oscilador en 3D, o como que cada átomo son 3 osciladores 1D independientes. Lo haremos como 3 osciladores 1D independientes. Si la frecuencia de estos osciladores está dada por $\nu$

Frecuencia de los osciladores $=  \nu$

Entonces, las posibles energías de cada oscilador serán (según cuántica): 

$$\epsilon_n = (n+1/2)h\nu$$

donde $n$ es un número natural. 

Este es un problema discreto, así que está "más fácil". Para cada número cuántico (estado) hay una sola posibilidad, porque no hay estados degenerados. Entonces: 

$$Z(T,V, 1) = \sum_{n = 0}^\infty e^{-(n+1/2)h\nu /k_B T}$$

Si hubiera alguna degeneración por los espines, simplemente hay que multiplicar esto por la degeneración, que sería el número de estados. 

Ahora, obtenemos $p_n$ y de ahí obtenemos $\langle E \rangle$. 



$$p_n = \frac{ e^{-(n+1/2)h\nu /k_B T}}{\sum_{n = 0}^\infty e^{-(n+1/2)h\nu /k_B T}}$$

$$\langle E \rangle = 3N \frac{ \sum _{n= 0}^\infty  (n+1/2)h \nu e^{-(n+1/2)h\nu /k_B T}}{\sum_{n = 0}^\infty e^{-(n+1/2)h\nu /k_B T}}$$

Ahora reducimos esto. Talacha. Notemos que: 

 $$\langle E \rangle = 3Nk_B T^2 \frac{\partial }{\partial T}\log(\sum (n+1/2)h e^{-(n+1/2)h\nu /k_B T})$$
 
 $$ = 3/2 N h \nu + 3Nk_B T^2 \frac{\partial }{\partial T}\log(\sum e^{-nh\nu /k_B T})$$

$$ = \frac{3}{2}N h\nu - 3Nk_B T^2 \frac{\partial}{\partial T}\log((1-e^{-h\nu/k_B T})^{-1})$$

$$ = \frac{3}{2}N h\nu - 3Nk_B T^2 \frac{h\nu}{1-e^{-h\nu/k_B T}}$$

De aquí que el calor específico (después de un poco de talacha salga: 

$$C_V = \frac{\partial}{\partial T} \langle E \rangle = 3Nk_B \Big(\frac{h\nu}{k_B T}\Big)^2 \frac{e^{-h\nu/k_B T}}{(1-e^{-h\nu/k_B T})^2}$$

El modelo es muy impresiso, de todas formas, rompió un paradigma. Hasta 1907, se creía que la capacidad calorífica de los sólidos cristalinos era constante y era $3N k_B$. Para temperaturas altas, este límite se puede obtener por series de Taylor. La exponencial se va a $1$, pero 1-exponencial se va a 0. Si hacemos la serie de Tylor, tenemos que arriba es 1+ algo, abajo es $(x+x^2/2+\dots)^2$, que quedandonos con el primer térmono no sale $1/x^2$ todo multiplicado por $x^2$, hace que sólo quede $3Nk_B$. Para temperaturas bajas, sin embargo, el límite se va a 0. Eso llevó a experimentos a temperaturas bajas, donde confiraron este efecto. No decaia como esta fórmula, pero parecido. 

## Distribución de Maxwell

Ahora vamos a estudiar la distribución de Maxwell. Para eso pensemos en el caso clássico de un gas (como lo hizo Maxwell). Las partículas en ese caso son distinguibles. No nos importa realmente, pero ahora nos convendrá no cargar con el $N!$. Entonces: 

$$Z = \int \rho(\vec{x}, \vec{p}) e^{-\beta H(\vec{x}, \vec{p})}d\vec{x}d\vec{p}$$  

pero ya habíamos visto que la $\rho(\vec{p}) = 1$. Por lo tanto la probabilidad de tener una partícicula en un estado es: 

$$p_\epsilon(\vec{x},\vec{p})d\vec{p}d\vec{x} = \frac{ e^{-\beta (p^2/2m + U(\vec{r}))} d\vec{p}d\vec{x}}{\int_V d\vec{x}\int\int\int_{-\infty}^\infty e^{-\beta (p^2/2m + U(\vec{r}))} d\vec{p}}$$

En ese paso justamente es donde eliminaríamos el $N!$. Como $N!$ sería justamente la degeneración del sistema, tendríamos que multiplicar por $N!$ y ahí se va. De aquí que la densidad de partículas sea: 

$$n(\vec{x},\vec{p})d\vec{p}d\vec{x} = \frac{N e^{-\beta (p^2/2m + U(\vec{r}))} d\vec{p}d\vec{x}}{\int_V d\vec{x}\int\int\int_{-\infty}^\infty e^{-\beta (p^2/2m + U(\vec{r}))} d\vec{p}}$$

Esta formulita les servirá para todo lo que sean gases. Por supuesto, de aquí sale la distribución de partículas con una cierta energía y con una cierta velocidad. Si no hay un potencial $U$, esto se reduce bastante, pues podemos integrar todo con respecto de las posiciones y sacar las 3 integrales de abajo, lo que nos queda:  

$$n(\vec{p})d\vec{p} = \frac{N e^{-\beta p^2/2m} d\vec{p}}{(2\pi m k_B T)^{3/2}}$$

que es una distribución gaussiana para cada una de las direcciones de $p$. Ya habíamos visto que si este era el caso, entonces la distribución de velocidades seguía una distribución de Maxwell, pero ahora tenemos exactamente qué distribución. Sustituyendo en lo que teníamos de una clase vieja: 

$$n(v)dv = \frac{\sqrt{2}}{\sqrt{\pi} \sigma^3} v^2 \exp\Big(\frac{-v}{2\sigma^2}\Big)dv$$

Ahora reconocemos que $\sigma = (2k_B T/m)^{1/2}$ si hacemos el cambio de variable de $p = mv$. 

$$ n(v)dv = \frac{ N m^{3/2}}{(2\pi k_B T)^{3/2}} e^{-mv^2/(2k_B T)} v^2 dv$$

Otro bonito problema es pensar en la distribución de partículas en un volumen, con independencia de la velocidad de las partículas. Si no hay potencial oviamente eso sólo dará una densidad constante, pero si ponemos un potencial externo, por ejemplo gravedad o un potencial centrífugo, podemos tener una densidad diferente según la posición. 

El primer paso es obtener $n(\vec{x})d\vec{x}$. Para eso integramos la parte de los momentos: 


$$n(\vec{x})d\vec{x} =\int\int\int_{-\infty}^\infty \frac{N e^{-\beta (p^2/2m + U(\vec{r}))} d\vec{x}}{\int_V d\vec{x}\int\int\int_{-\infty}^\infty e^{-\beta (p^2/2m + U(\vec{r}))} d\vec{p}} d\vec{p}$$

$$= \frac{Ne^{-\beta U(r)} d\vec{x}}{\int_V e^{-\beta U(r)}d\vec{x}}$$



Pensemos en el caso más sencillo:


$$U(h) = mgh$$

$$n(h)dh = \frac{Ne^{-\beta mgh} dh}{\int_V e^{-\beta mgh}dh} = \frac{Ne^{-\beta mgh} dh}{k_B T/mg}$$

$$= \frac{Nmg}{k_BT}e^{-mgh/k_B T}$$

Que es la  densidad de partículas como función de la altura!! Está padre, no? Si les gusta la astronomía, pueden imaginar que esta clase de cálculo sirve para estimar la densidad de partículas en por ejemplo, una nebulosa, etc. 

 ## Gas diatómico

El último ejemplo que quiero revisar es el del gas diatómico. Aquí regresaré a la versión cuántica, porque, igual que con el gas ideal, es más natural hacer los conteos. Debo aclarar que eso no convierte del todo el problema en un problema cuántico, porque en realidad trabajaremos en los límite donde no importa si se trata de un ensamble clásico o cuántico, es decir, donde aún no se ven los efectos "atractivos" o "repulsivos" de las partículas por el principio de incertidumbre. Dicho esto, como las partículas serán indistinguibles: 

$$Z(T,V,N) = \frac{Z(T,V,1)^N }{N!}$$

Insisto, esto es porque las partículas son ideales e indistinguibles. Aunque sean diatómicas, consideraremos que hay una muy baja densidad, como para que interactuen entre ellas. 

Ahora bien, tanto clásicamente, como cuánticamente, podemos separar la energía de un objeto en la suma de su energía traslacional, su energía rotacional y su energía interna. Por energía interna me refiero a cualquier otra producida "dentro" de la partícula, por ejemplo la vibracional si pensamos en que la molécula se forma por dos partículas unidas por un resrte. También podría estar ahí su energía eléctrica, nuclear, etc. Vamos a considerar el caso sencillo donde la única energía interna es la rotacional. 

$$Z(T,V,1) = \sum_i e^{-\beta E_i} = \sum_i \sum_j e^{-\beta (E_{t_i} + E_{r_j})} = (\sum_i e^{-\beta E_{t_i}}) (\sum_j e^{-\beta E_{r_j}})$$

Aquí la primera suma va sobre los momentos, la segunda sobre los momentos angulaes. La primera parte es la que ya teníamos, peus es simplemente el gas ideal: 

$$Z(T,V,1) = V\Big(\frac{2\pi m k_B T}{h^2}\Big)^{3/2} (\sum_j e^{-\beta E_{r_j}})$$

Así que sólo tenemos que calcular la suma para las energías rotacionales.
