```julia
N = 1  # tamao de la poblacin

i = N * 0.8  # proporcin de infectados (respecto de la poblacin)
r =  N * 0.01  # proporcin de recuperados (respecto de la poblacin)

p = (1/1000, 2, 0.5, N)  # m, b, g, N

u0 = [N - r, i, r]  # el orden es "S I R"

tspan = (0.0, 3000)

problema = ODEProblem(sir!, u0, tspan, p)

sol_tot = solve(problema, PFRK87(), adaptive=false, dt = 0.1)

#as se accede a la solucin obtenida
sol_u = sol_tot.u

# y grafiquemos como

t_val = sol_tot.t 

sol_S = [u[1] for u in sol_tot.u]
sol_I = [u[2] for u in sol_tot.u]
sol_R = [u[3] for u in sol_tot.u]

plot(sol_S, sol_I, label = "S v.s. I")

```
